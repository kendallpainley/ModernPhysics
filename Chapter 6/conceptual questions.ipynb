## 6.1 Blackbody Radiation

**Q1. Which surface has a higher temperature — the surface of a yellow star or that of a red star?**

A yellow star has a higher surface temperature. According to Wien's displacement law, the peak emission wavelength is inversely proportional to temperature ($\lambda_max · T$ = constant). Yellow light has a shorter wavelength than red light, so the yellow star's peak emission occurs at a shorter wavelength, indicating a higher surface temperature.

---

**Q2. Describe what you would see when looking at a body whose temperature is increased from 1000 K to 1,000,000 K.**

At $1000 K$, the body glows a dull red, as its peak emission lies in the infrared but the tail of the distribution just enters the red visible range. As temperature increases, the peak wavelength shifts to shorter values. The glow moves through bright red, orange, yellow, and white as more of the visible spectrum is covered. Above $6000–10,000 K$, the object appears bright white or blue-white as the peak moves into the visible and then into the ultraviolet. At $1,000,000 K$, the peak emission lies in the UV/X-ray range; the body would radiate intense ultraviolet and X-ray radiation and would appear blindingly bright blue-white in any visible component.

---

**Q3. Explain the color changes in a hot body as its temperature is increased.**

As temperature rises, Wien's displacement law says that the peak wavelength of emitted radiation decreases. At low temperatures the emission peak is in the infrared and only the long-wavelength tail of the distribution falls in the visible range, giving a red glow. As temperature increases, the peak moves through red, orange, yellow, and into the visible range more, increasing the glow. Simultaneously, Stefan's law ($P ∝ T⁴$) means total radiated power increases drastically, so the body also becomes much brighter. Eventually, at very high temperatures, all visible wavelengths are emitted strongly and the object appears white.

---

**Q4. Speculate as to why UV light causes sunburn, whereas visible light does not.**

UV photons have shorter wavelengths and therefore higher energies ($E = hf = hc/λ$) than visible light photons. The energy of a UV photon is enough to break chemical bonds in biological molecules and to ionize atoms within skin cells, causing cellular damage. Visible light photons carry less energy per photon and cannot break these bonds, so they are absorbed as heat.

---

**Q5. Two cavity radiators are constructed with walls made of different metals. At the same temperature, how would their radiation spectra differ?**

They would not differ. A blackbody emits radiation that depends only on temperature. The cavity radiator is an idealized perfect absorber and emitter. Both cavities at the same temperature $T$ would produce identical blackbody radiation spectra described by Planck's law, with the same peak wavelength and the same total emitted power.

---

**Q6. Discuss why some bodies appear black, other bodies appear red, and still other bodies appear white.**

The color of a body is determined by which wavelengths of incident light it absorbs versus reflects. A black body absorbs all visible wavelengths and reflects none, so no light reaches the eye. A red body absorbs most visible wavelengths but reflects primarily red wavelengths, which reach the eye. A white body reflects all visible wavelengths equally and absorbs very little, so the full spectrum of reflected light reaches the eye and it appears white.

---

**Q7. If everything radiates electromagnetic energy, why can we not see objects at room temperature in a dark room?**

At room temperature, Wien's displacement law gives a peak emission wavelength of about $10 µm$, which is infrared. The human eye detects only wavelengths from 400 to 700 nm. Room-temperature objects emit little power in this range; basically all their radiated energy is in the infrared.

---

**Q8. How much does the power radiated by a blackbody increase when its temperature (in K) is tripled?**

From Stefan's law, $P = σAT⁴$. If temperature is tripled ($T → 3T$):

$$P' = \sigma A (3T)^4 = 81\,\sigma A T^4 = 81\,P$$

The radiated power increases by a factor of **81**.


## 6.2 Photoelectric Effect

**Q9. For the same monochromatic light source, would the photoelectric effect occur for all metals?**

No. Whether the photoelectric effect occurs depends on whether the photon energy exceeds the work function of the particular metal. Different metals have different work functions. For a fixed light source, the effect will occur for metals with low enough work functions but not for metals with high work functions.

---

**Q10. In the interpretation of the photoelectric effect, how is it known that an electron does not absorb more than one photon?**

The evidence comes from stopping potential measurements. If electrons could accumulate energy from multiple photons, the maximum kinetic energy would depend on the light intensity (more photons arriving per unit time would give electrons more opportunities to absorb multiple photons). But experiments show that the stopping potential is independent of intensity; it depends only on frequency.

---

**Q11. Explain how you can determine the work function from a plot of stopping potential versus frequency. Can you also determine Planck's constant?**

From Einstein's photoelectric equation, $K_max = hf − ϕ$, and since $K_max = eΔV_s$:

$$e\Delta V_s = hf - \phi$$

A plot of stopping potential ($ΔV_s$) versus frequency ($f$) is a straight line. The slope equals $h/e$, from which Planck's constant h can be determined. The y-intercept is $−ϕ/e$, from which the work function $ϕ$ can be read. The x-intercept gives the cut-off frequency $f_c = ϕ/h$.

---

**Q12. Suppose you plot detected current versus applied potential difference in a photoelectric experiment. What information is obtained? Can you determine Planck's constant or the work function?**

- The **saturation current** (the plateau at positive potential) is proportional to the number of photoelectrons emitted per unit time, and therefore proportional to the intensity of the incident light.
- The **stopping potential** (negative voltage at which current drops to zero) gives the maximum kinetic energy of the photoelectrons via K_max = eΔV_s.
- **Planck's constant cannot be determined** from a single I-V curve alone, because that would require knowing the photon frequency and work function independently.
- The **work function cannot be determined** from this plot alone either — you would need to know the frequency of the incident radiation and then use K_max = hf − ϕ.
- However, by repeating the experiment at several frequencies and plotting stopping potential vs. frequency, both h and ϕ can be found (as described in Q11).

---

**Q13. Speculate how increasing the temperature of a photoelectrode affects the outcomes of the photoelectric effect experiment.**

At higher temperatures, the conduction electrons in the metal have greater thermal kinetic energy. This means some electrons already have more energy before the photon interaction occurs. These thermally excited electrons require less photon energy to escape the surface. As a result, the measured stopping potential would increase slightly (some electrons are emitted with more kinetic energy), and the effective cut-off frequency would decrease slightly (lower-energy photons could liberate already thermally excited electrons). The overall photocurrent may also increase at sufficiently high temperatures due to thermionic emission — spontaneous electron emission from heat alone, separate from the photoelectric effect.

---

**Q14. Which aspects of the photoelectric effect cannot be explained by classical physics?**

Three key aspects defy classical explanation:

1. **No lag time:** Electrons are emitted essentially instantaneously even at very low intensities. Classical wave theory predicts it would take significant time for electrons to accumulate enough energy from a low-intensity wave.
2. **Kinetic energy independent of intensity:** Classical theory predicts that a more intense wave should give electrons more energy. Instead, maximum kinetic energy depends only on frequency, not on intensity.
3. **Cut-off frequency:** Below a certain frequency, no electrons are emitted regardless of intensity. Classical theory predicts emission should occur at any frequency given sufficient intensity.

---

**Q15. Is the photoelectric effect a consequence of the wave character or the particle character of radiation?**

The photoelectric effect is a consequence of the **particle character** of radiation. It can only be explained by treating light as discrete photons, each carrying energy E = hf. The one-on-one interaction between a single photon and a single electron — with an all-or-nothing energy transfer — is inherently particle-like behavior. Wave models, which treat energy as continuously distributed across a wavefront, cannot account for the instantaneous emission, the frequency threshold, or the independence of kinetic energy from intensity.

---

**Q16. Sodium, iron, and molybdenum have work functions 2.5 eV, 3.9 eV, and 4.2 eV respectively. Which will emit photoelectrons when illuminated with 400-nm light?**

First, calculate the photon energy:

$$E = \frac{hc}{\lambda} = \frac{1240 \text{ eV·nm}}{400 \text{ nm}} = 3.1 \text{ eV}$$

Comparing to work functions:
- **Sodium** (ϕ = 2.5 eV): 3.1 eV > 2.5 eV → **will emit photoelectrons** ✓
- **Iron** (ϕ = 3.9 eV): 3.1 eV < 3.9 eV → will not emit ✗
- **Molybdenum** (ϕ = 4.2 eV): 3.1 eV < 4.2 eV → will not emit ✗

Only **sodium** will exhibit the photoelectric effect under 400-nm illumination.


## 6.3 The Compton Effect

**Q17. Discuss similarities and differences between the photoelectric effect and the Compton effect.**

|                | Photoelectric Effect                      | Compton Effect                                          |
|----------------|-------------------------------------------|---------------------------------------------------------|
| **Similarity** | Both treat radiation as discrete photons  | Both treat radiation as discrete photons                |
| **Similarity** | Both involve photon–electron interactions | Both involve photon–electron interactions               |
| **Difference** | Photon is completely absorbed             | Photon is scattered (survives with less energy)         |
| **Difference** | Electron is bound in a metal              | Electron is effectively free (loosely bound)            |
| **Difference** | Photon energy ~ eV range (UV/visible)     | Photon energy ~ keV range (X-rays)                      |
| **Difference** | Electron is ejected from the material     | Electron recoils; photon continues at longer wavelength |

Both effects were critical in establishing the particle nature of light and cannot be explained by classical wave theory.

---

**Q18. Which has greater momentum: a UV photon or an IR photon?**

A **UV photon** has greater momentum. The momentum of a photon is p = h/λ. UV photons have shorter wavelengths than IR photons, so they have larger momentum. Equivalently, UV photons have higher frequency and therefore higher energy (E = hf), and since p = E/c, higher energy means greater momentum.

---

**Q19. Does changing the intensity of a monochromatic light beam affect the momentum of individual photons? Does it affect the net momentum of the beam?**

- **Individual photon momentum:** No. The momentum of each photon is p = h/λ, which depends only on wavelength. Changing intensity means changing the number of photons per unit time, not the energy or momentum of each individual photon.
- **Net beam momentum:** Yes. Higher intensity means more photons pass a given area per unit time, so the total momentum transported by the beam per unit time (i.e., the radiation pressure force) increases proportionally with intensity.

---

**Q20. Can the Compton effect occur with visible light? If so, will it be detectable?**

Yes, the Compton effect can in principle occur with visible light — there is no fundamental barrier to photon-electron scattering at any wavelength. However, it is effectively **undetectable** with visible light. The Compton wavelength shift is fixed at Δλ = λ_c(1 − cosθ), where λ_c = 2.43 pm. For visible light (wavelengths ~400–700 nm), the fractional shift Δλ/λ is of order 10⁻⁵ or smaller — far too tiny to measure with any practical spectrometer. The effect is only clearly observable when the incident wavelength is comparable to the Compton wavelength, which is why X-rays (λ ~ 0.01–0.1 nm) are used.

---

**Q21. Is it possible in the Compton experiment to observe scattered X-rays with a shorter wavelength than the incident radiation?**

No. In Compton scattering, the incident photon transfers some of its energy and momentum to the recoiling electron. The photon always loses energy (except at θ = 0°, where there is no interaction). Since E = hc/λ, a loss of energy corresponds to an increase in wavelength. The scattered photon always has a wavelength equal to or greater than the incident wavelength. A shorter wavelength would imply the photon gained energy, which would violate conservation of energy for this system (the electron starts at rest and cannot supply energy to the photon).

---

**Q22. Show that the Compton wavelength has the dimension of length.**

The Compton wavelength is defined as:

$$\lambda_c = \frac{h}{m_0 c}$$

Checking dimensions:

$$[\lambda_c] = \frac{[h]}{[m_0][c]} = \frac{\text{J·s}}{\text{kg} \cdot \text{m/s}} = \frac{\text{kg·m}^2\text{/s}^2 \cdot \text{s}}{\text{kg·m/s}} = \frac{\text{kg·m}^2}{\text{kg·m}} = \text{m}$$

The Compton wavelength has dimensions of **length**. ✓

---

**Q23. At what scattering angle is the wavelength shift equal to the Compton wavelength?**

The Compton shift is Δλ = λ_c(1 − cosθ). Setting Δλ = λ_c:

$$\lambda_c = \lambda_c(1 - \cos\theta)$$
$$1 = 1 - \cos\theta$$
$$\cos\theta = 0 \implies \theta = 90°$$

The wavelength shift equals the Compton wavelength at a scattering angle of **90°**.


## 6.4 Bohr's Model of the Hydrogen Atom

**Q24. Explain why the patterns of bright emission lines and dark absorption lines coincide for a given gaseous element.**

Both emission and absorption involve the same discrete energy level transitions. In emission, an electron drops from a higher to a lower energy level, releasing a photon of exactly the energy difference. In absorption, an electron in a lower level absorbs a photon of exactly the right energy to jump to a higher level. Since the energy levels are fixed by the atomic structure, the photon energies — and therefore wavelengths — involved are identical in both processes. Kirchhoff and Bunsen confirmed experimentally that the bright lines in the emission spectrum of any element fall at exactly the same spectral positions as the dark lines in its absorption spectrum.

---

**Q25. Do the various spectral lines of the hydrogen atom overlap?**

The different series (Lyman, Balmer, Paschen, etc.) do not overlap with each other in terms of wavelength ranges, because each series converges to a distinct series limit. Within a single series, lines get progressively closer together as n_i → ∞, converging to the series limit but never crossing into another series' range. The series lie in completely different spectral regions: Lyman in the UV, Balmer partly in the visible and partly in the UV, Paschen in the near infrared, and so on. The series do **not** overlap.

---

**Q26. The Balmer series was discovered before the Lyman or Paschen series. Why?**

The Balmer series (n_f = 2) includes lines in the visible portion of the spectrum (the four visible lines at 656, 486, 434, and 410 nm). In the nineteenth century, spectroscopic instruments were primarily designed to study visible light. The Lyman series lies entirely in the ultraviolet and the Paschen series lies entirely in the infrared — both regions required specialized equipment not widely available until the twentieth century. Because the Balmer series was directly visible to the eye through a prism or diffraction grating, it was the most accessible and was discovered first (by Johann Balmer in 1885).

---

**Q27. When hydrogen's absorption spectrum is analyzed at room temperature, Lyman series lines are found but no Balmer series lines. What does this tell us?**

The Balmer series involves transitions from the n = 2 level upward (absorption into n = 3, 4, 5, ...). For Balmer absorption lines to appear, a significant number of hydrogen atoms must have their electrons in the n = 2 state. At room temperature, the thermal energy (k_B T ≈ 0.025 eV) is far smaller than the energy gap between n = 1 and n = 2 (~10.2 eV). By Boltzmann statistics, essentially all hydrogen atoms at room temperature have their electrons in the ground state (n = 1). Therefore, only Lyman series absorption (from n = 1 upward) is observed. The absence of Balmer absorption lines confirms that **virtually all hydrogen atoms at room temperature are in the ground state**.

---

**Q28. Why are hydrogen absorption lines strongest in stars at ~9000 K, weaker in the sun, and nearly absent in very hot or very cool stars?**

The strength of hydrogen absorption lines depends on the fraction of hydrogen atoms with electrons in the n = 2 state (for Balmer lines) at the stellar surface temperature:

- **Very cool stars (< 3500 K):** Thermal energy is insufficient to excite electrons from n = 1 to n = 2. Nearly all hydrogen is in the ground state, so Balmer absorption (which requires n = 2 as the starting level) is very weak.
- **~9000 K:** This temperature provides just enough thermal energy to populate the n = 2 level significantly, maximizing the number of atoms that can absorb Balmer photons. Absorption lines are at their strongest.
- **Sun (~6000 K):** Fewer atoms are thermally excited to n = 2 compared to 9000 K, so lines are somewhat weaker.
- **Very hot stars (> 25,000 K):** At these temperatures, hydrogen atoms are largely ionized — the electron has been stripped completely from the proton. With no bound electrons, there are no atomic transitions and no absorption lines.

---

**Q29. Discuss the similarities and differences between Thomson's model and Bohr's model of the hydrogen atom.**

|                | Thomson's Model                                     | Bohr's Model                                      |
|----------------|-----------------------------------------------------|---------------------------------------------------|
| **Similarity** | Contains electrons                                  | Contains electrons                                |
| **Similarity** | Atom is electrically neutral overall                | Atom is electrically neutral overall              |
| **Difference** | Positive charge is a diffuse cloud filling the atom | Positive charge concentrated in a tiny nucleus    |
| **Difference** | Electrons embedded within the positive cloud        | Electrons orbit the nucleus at specific distances |
| **Difference** | No quantization of energy or orbits                 | Quantized orbits and energy levels                |
| **Difference** | Cannot explain emission spectra                     | Correctly predicts hydrogen emission spectrum     |

---

**Q30. Discuss the way in which Thomson's model is nonphysical. Support with experimental evidence.**

Thomson's "plum pudding" model proposed that positive charge is spread uniformly throughout the atom, with electrons embedded within it. The Rutherford gold foil experiment (1909) showed this to be untenable. When alpha particles were fired at a thin gold foil, most passed straight through, but a small fraction were deflected at large angles — some even bouncing almost straight back. This is impossible in Thomson's model: a diffuse positive charge cannot exert a strong enough concentrated force to deflect a heavy, fast alpha particle significantly. The large-angle scattering can only be explained if nearly all the positive charge and mass are concentrated in an extremely small, dense region — the nucleus. Thomson's model is therefore inconsistent with the experimental evidence.

---

**Q31. If an electron in a hydrogen atom moves to an orbit with a larger radius, does the energy of the hydrogen atom increase or decrease?**

The energy **increases** (becomes less negative). From Bohr's model, the total energy of the electron in the nth orbit is E_n = −E_0/n², where E_0 = 13.6 eV. Larger orbits correspond to larger n, and as n increases, 1/n² decreases, making E_n less negative (closer to zero). The atom has more energy in a larger orbit. This is consistent with the fact that energy must be *supplied* (absorbed) to move an electron to a higher orbit, and energy is *released* (emitted) when it falls to a lower orbit.

---

**Q32. How is energy conserved when an atom makes a transition from a higher to a lower energy state?**

When an electron transitions from a higher energy orbit (E_n) to a lower energy orbit (E_m, where E_m < E_n), the difference in energy is carried away by an emitted photon. The photon's energy is exactly:

$$E_{photon} = hf = E_n - E_m$$

The total energy of the system (atom + emitted photon) remains constant. The atom loses exactly as much energy as the photon gains. This is Bohr's second quantization condition and directly explains the discrete spectral lines of hydrogen.

---

**Q33. For a transition from the (n+1)th to the nth orbit, is the wavelength of the emitted photon longer for larger or smaller values of n?**

The wavelength is **longer for larger values of n**. The energy of the emitted photon is:

$$\Delta E = E_0\left(\frac{1}{n^2} - \frac{1}{(n+1)^2}\right)$$

For large n, adjacent orbits are very close in energy: 1/n² − 1/(n+1)² ≈ 2/n³, which decreases as n increases. A smaller energy difference means a lower-frequency (longer-wavelength) photon. Physically, at large n the energy levels are closely spaced and transitions between them release very little energy, corresponding to long-wavelength (microwave/radio) photons.

---

**Q34. Discuss why the allowed energies of the hydrogen atom are negative.**

The energy is defined relative to the reference state of zero total energy, which corresponds to an electron at rest at infinite distance from the nucleus (completely ionized). A bound electron must have less energy than this reference — meaning it has negative total energy. The negative sign signifies that the electron is bound and cannot escape spontaneously; energy must be *supplied* to free it. The magnitude of the energy represents the binding energy. As n → ∞, E_n → 0, corresponding to the electron becoming unbound (ionization limit). All finite-n states, being bound states, therefore have negative energy.

---

**Q35. Can a hydrogen atom absorb a photon whose energy is greater than 13.6 eV?**

Yes. The ionization energy of hydrogen from the ground state is 13.6 eV. A photon with exactly 13.6 eV would just barely ionize the atom, leaving the freed electron with zero kinetic energy. A photon with energy greater than 13.6 eV can also be absorbed — the atom is ionized, and the excess energy above 13.6 eV is carried away as kinetic energy of the freed electron. This is the photoelectric effect at the atomic level. There is no upper limit; any photon with energy ≥ 13.6 eV can ionize the ground-state hydrogen atom.

---

**Q36. Why can you see through glass but not through wood?**

The transparency of a material depends on whether its electrons can absorb the photons passing through it. In glass, the energy gaps between electronic states are larger than the energies of visible photons — visible light photons do not have enough energy to excite electrons across the gap, so they pass through without being absorbed. In wood, the complex molecular and cellular structure contains many energy states at spacings corresponding to visible photon energies, allowing strong absorption. Additionally, wood contains numerous surfaces, fibers, and pigments that scatter and absorb light at every wavelength.

---

**Q37. Do gravitational forces have a significant effect on atomic energy levels?**

No. Gravitational forces between the electron and proton in a hydrogen atom are many orders of magnitude weaker than the electrostatic (Coulomb) force. The ratio of gravitational to electrostatic force between an electron and a proton is:

$$\frac{F_{grav}}{F_{elec}} = \frac{Gm_e m_p}{ke^2} \approx 4 \times 10^{-40}$$

This is so vanishingly small that gravity plays absolutely no role in determining atomic structure, energy levels, or spectral lines. Electromagnetism completely dominates at the atomic scale.

---

**Q38. Show that Planck's constant has the dimensions of angular momentum.**

Angular momentum has dimensions [L] = kg·m²/s.

Planck's constant h appears in E = hf, so:

$$[h] = \frac{[E]}{[f]} = \frac{\text{J}}{\text{Hz}} = \frac{\text{kg·m}^2\text{/s}^2}{\text{1/s}} = \text{kg·m}^2\text{/s}$$

This is the same as the dimensions of angular momentum (mass × length² / time = kg·m²/s). ✓

This is not a coincidence — Bohr's quantization condition L_n = nℏ directly expresses angular momentum in units of the reduced Planck's constant.


## 6.5 De Broglie's Matter Waves

**Q39. Which type of radiation is most suitable for observing diffraction patterns on crystalline solids: radio waves, visible light, or X-rays? Explain.**

**X-rays** are most suitable. For diffraction to occur, the wavelength of the radiation must be comparable to the spacing between the crystal planes being probed (typically ~0.1–1 nm for atomic lattice spacings). X-rays have wavelengths in the range of 0.01–10 nm, which matches crystal plane spacings well. Radio waves (~1 mm–1 km) and visible light (~400–700 nm) have wavelengths many orders of magnitude too large to be diffracted by atomic-scale crystal planes; they would simply pass through or reflect without producing a usable diffraction pattern.

---

**Q40. How would diffraction patterns from a typical crystal be affected if γ-rays were used instead of X-rays?**

Gamma rays have wavelengths much shorter than X-rays (typically < 0.01 nm), which are much smaller than typical crystal plane spacings (~0.1–1 nm). Because Bragg's law requires nλ = 2d sinθ, a very small λ compared to d would require very small scattering angles θ to satisfy this condition. The diffraction peaks would be compressed to very small angles near the forward direction, making them extremely difficult to resolve and distinguish experimentally. In practice, the diffraction pattern would be much harder to observe and interpret compared to using X-rays.

---

**Q41. If an electron and a proton are traveling at the same speed, which has the shorter de Broglie wavelength?**

The **proton** has the shorter de Broglie wavelength. From λ = h/p = h/(mv), at the same speed v, the wavelength is inversely proportional to mass. Since the proton is approximately 1836 times more massive than the electron, its momentum at the same speed is 1836 times greater, giving a wavelength 1836 times shorter than that of the electron.

---

**Q42. If a particle is accelerating, how does this affect its de Broglie wavelength?**

Acceleration changes the particle's speed and therefore its momentum. Since λ = h/p, an increase in momentum (from speeding up) causes the de Broglie wavelength to **decrease**, and a decrease in momentum (from slowing down) causes the wavelength to **increase**. The wavelength changes continuously as the particle's momentum changes — there is no single fixed wavelength for an accelerating particle.

---

**Q43. Why is the wave-like nature of matter not observed every day for macroscopic objects?**

The de Broglie wavelength λ = h/p = h/(mv) is inversely proportional to momentum. For macroscopic objects, the mass is so large that even at low speeds the momentum is enormous compared to Planck's constant (h ≈ 6.63 × 10⁻³⁴ J·s). This makes the de Broglie wavelength immeasurably small — far smaller than any atomic nucleus. For example, a 0.1-kg ball moving at 1 m/s has λ ≈ 6.6 × 10⁻³³ m. No physical apparatus can detect interference or diffraction at such scales. Wave behavior is only observable when the wavelength is comparable to the relevant length scales of the system (e.g., atomic spacings for electron diffraction).

---

**Q44. What is the wavelength of a neutron at rest? Explain.**

A neutron at rest has zero momentum (p = 0). Since λ = h/p, this gives an infinite wavelength — physically interpreted to mean that the neutron has no defined de Broglie wavelength in the traditional sense. A particle at rest would correspond to a plane wave spread uniformly over all space, consistent with infinite wavelength. In practice, a neutron can never be perfectly at rest; the Heisenberg uncertainty principle prevents a particle from simultaneously having precisely zero momentum and a defined position.

---

**Q45. Why does the Davisson–Germer experiment need to be enclosed in a vacuum chamber? What would happen if it were not evacuated?**

The experiment must be in a vacuum to prevent electrons in the beam from colliding with air molecules. Such collisions would scatter the electrons randomly, changing their direction and kinetic energy unpredictably. This would broaden and wash out the sharp diffraction peaks, or destroy the coherent beam entirely, making it impossible to observe the clean interference pattern required to confirm matter-wave behavior. In vacuum, electrons travel unimpeded from the gun to the nickel target and then to the detector, preserving both their wavelength and their directional coherence.


## 6.6 Wave-Particle Duality

**Q46. Give an example of an experiment in which light behaves as waves. Give an example in which light behaves as a stream of photons.**

- **Wave behavior:** Young's double-slit experiment. Light passing through two closely spaced slits creates a pattern of alternating bright and dark fringes on a screen, which can only be explained by wave interference — the overlapping of crests and troughs of the two emerging waves.
- **Particle behavior:** The photoelectric effect. Light striking a metal surface ejects electrons only when the photon frequency exceeds a threshold, with each electron receiving the energy of exactly one photon. This all-or-nothing energy transfer, independent of intensity, cannot be explained by a continuous wave and requires treating light as discrete photon particles.

---

**Q47. How does the interference of water waves differ from the interference of electrons? How are they analogous?**

**Differences:**
- Water wave interference is a classical phenomenon involving the superposition of pressure oscillations in a physical medium (water). It can be fully described without quantum theory.
- Electron interference involves individual quantum particles. Each electron is detected at a definite point, yet after many electrons accumulate, an interference pattern builds up. There is no classical medium "waving" — the wave is a probability amplitude.

**Analogies:**
- Both produce the same mathematical pattern of constructive and destructive interference, described by the same wave equations.
- Both show bright maxima where path differences are integer multiples of the wavelength, and minima where path differences are half-integer multiples.
- In both cases, the pattern depends on wavelength, slit spacing, and screen distance in the same way (dsinθ = nλ).

---

**Q48. Give at least one argument in support of the matter-wave hypothesis.**

The Davisson-Germer experiment (1927) provides the most direct evidence. When a beam of electrons was directed at a crystalline nickel surface, a distinct diffraction pattern was observed — an intensity maximum appeared at a specific scattering angle. This pattern is characteristic of wave diffraction and is analogous to the X-ray diffraction patterns produced by the same crystal. The measured electron wavelength (1.64 Å, derived from the diffraction geometry) matched precisely the de Broglie wavelength predicted from the electrons' momentum (1.67 Å). This quantitative agreement can only be explained if electrons have an associated wave with the wavelength λ = h/p.

---

**Q49. Give at least one argument in support of the particle nature of radiation.**

The Compton effect provides compelling evidence. When X-rays scatter off loosely bound electrons in a material, the scattered X-rays have a longer wavelength than the incident X-rays, and the wavelength shift depends precisely on the scattering angle according to Δλ = (h/m₀c)(1 − cosθ). This formula was derived by treating the X-ray photon as a relativistic particle with momentum p = h/λ undergoing a billiard-ball-type elastic collision with an electron. Classical wave theory predicts no wavelength shift at all in scattering. The perfect quantitative agreement between the particle-collision prediction and the experimental data is powerful evidence that radiation carries momentum and behaves as a stream of particles.

---

**Q50. Explain the importance of the Young double-slit experiment.**

Young's double-slit experiment (1803) was historically decisive in establishing the wave nature of light, overturning Newton's corpuscular (particle) theory. The interference fringes it produced can only be explained by wave superposition — particles fired through two slits would produce two bright bands, not an interference pattern. In the modern quantum context, the experiment takes on even deeper significance: when performed with single photons or single electrons arriving one at a time, an interference pattern still accumulates over many events. This demonstrates that the wave nature belongs to individual quantum particles, not to collective behavior. It embodies the central mystery of quantum mechanics — a single particle appears to "go through both slits" simultaneously, and any attempt to observe which slit it passed through destroys the interference pattern.

---

**Q51. Does the Heisenberg uncertainty principle allow a particle to be at rest in a designated region in space?**

No. The Heisenberg uncertainty principle states ΔxΔp ≥ ℏ/2. If a particle were at rest in a designated region, it would simultaneously have a precisely defined position (Δx = finite) and precisely defined momentum of zero (Δp = 0). But Δp = 0 would require Δx = ∞, which contradicts confinement to a specific region. Therefore, a particle cannot be simultaneously at rest and confined to a specific region. Any particle confined in space must have some spread in momentum — and therefore some kinetic energy — which is the quantum mechanical zero-point energy.

---

**Q52. Can the de Broglie wavelength of a particle be known exactly?**

No. From the Heisenberg uncertainty principle, exact knowledge of momentum (Δp = 0) would require complete uncertainty in position (Δx = ∞). Since λ = h/p, knowing λ exactly means knowing p exactly, which means the particle is completely delocalized over all space — its position is entirely undetermined. In practice, quantum particles always have some spread in wavelength (and thus momentum), described by a wave packet. A perfectly monochromatic matter wave with a single exact wavelength would be an infinite plane wave, which is not physically realizable.

---

**Q53. Do red light photons produce better resolution in a microscope than blue light photons?**

No — **blue light photons produce better resolution**. The resolving power of a microscope is fundamentally limited by the wavelength of the illuminating light (Rayleigh criterion: Δθ = 1.22λ/D). Shorter wavelengths allow finer detail to be resolved. Blue light has a shorter wavelength (~450 nm) than red light (~650 nm), so a microscope using blue light can resolve features roughly 30–40% smaller than one using red light. This is why electron microscopes — with de Broglie wavelengths thousands of times shorter than visible light — achieve vastly superior resolution.

**Q54. Discuss the main difference between an SEM and a TEM.**

| Feature                  | TEM (Transmission Electron Microscope)                          | SEM (Scanning Electron Microscope)                                   |
|--------------------------|-----------------------------------------------------------------|----------------------------------------------------------------------|
| **Electron interaction** | Beam passes *through* a very thin sample                        | Beam is scanned *over* the surface                                   |
| **Image source**         | Transmitted electrons                                           | Back-scattered electrons, secondary electrons, and emitted radiation |
| **Sample preparation**   | Must be ~100 nm thin; biological samples need chemical fixation | Bulk samples up to several cm; less demanding preparation            |
| **What is imaged**       | Internal structure, atomic-scale features                       | Surface topography                                                   |
| **Resolution**           | Can exceed 0.5 Å                                                | ~1 nm                                                                |
| **Magnification**        | Up to ~50 million times                                         | Up to ~250,000 times better than light microscope                    |
| **Beam energy**          | Up to ~400 keV                                                  | Up to ~40 keV                                                        |

Both exploit the wave nature of electrons (very short de Broglie wavelengths) to achieve resolution far beyond that of optical microscopes.
