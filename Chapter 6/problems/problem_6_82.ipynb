# Problem 6.82 — Wavelength and Energy of a Photon from Its Momentum

## Problem Statement

Find the wavelength and energy of a photon with momentum $p = 5.00 \times 10^{-29}$ kg·m/s.

---

## Background

For a photon, momentum and wavelength are related by de Broglie's relation:

$$p = \frac{h}{\lambda} \implies \lambda = \frac{h}{p}$$

And since $E = pc$ for a massless photon:

$$E = pc = \frac{hc}{\lambda}$$

---

## Strategy

1. Find $\lambda = h/p$
2. Find $E = pc$
3. Identify the spectral region from the wavelength


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── Given value ────────────────────────────────────────────────────
p = 5.00e-29    # photon momentum [kg·m/s]

# ── Wavelength ─────────────────────────────────────────────────────
lam   = h / p
E_J   = p * c
E_eV_v = E_J / eV

print("=" * 50)
print("Wavelength:  λ = h/p")
print("=" * 50)
print(f"  p   = {p:.2e} kg·m/s")
print(f"  h   = {h:.3e} J·s")
print(f"  λ   = h/p = {h:.3e} / {p:.2e}")
print(f"  λ   = {lam:.4e} m")
print(f"  λ   = {lam*1e6:.4f} µm  (infrared)")
print()
print("=" * 50)
print("Energy:  E = pc")
print("=" * 50)
print(f"  E   = {p:.2e} × {c:.3e}")
print(f"  E   = {E_J:.4e} J")
print(f"  E   = {E_eV_v:.6f} eV")
```

## Results

| Quantity | Value |
|----------|-------|
| Momentum $p$ | $5.00 \times 10^{-29}$ kg·m/s |
| Wavelength $\lambda$ | $1.325 \times 10^{-5}$ m = **13.3 µm** (mid-infrared) |
| Energy $E$ | $1.499 \times 10^{-20}$ J = **0.0936 eV** |

---

## Physical Interpretation

- At 13.3 µm, this photon is in the **mid-infrared** region — the type of radiation emitted by room-temperature objects.
- Its energy (~0.094 eV) is comparable to thermal energy at room temperature ($k_BT \approx 0.026$ eV), explaining why infrared photons cause heating.
- Such photons are far too low-energy to cause the photoelectric effect in any metal.


```python
# Context: where does this photon sit on the EM spectrum?
regions = {
    'X-ray\n(0.1 nm)':   0.1e-9,
    'UV\n(200 nm)':      200e-9,
    'Visible\n(500 nm)': 500e-9,
    'Near-IR\n(1 µm)':   1e-6,
    'Mid-IR\n(13 µm)':   lam,
    'Far-IR\n(100 µm)':  100e-6,
}
lam_vals  = list(regions.values())
lam_names = list(regions.keys())
energies  = [h*c/l/eV for l in lam_vals]
colors    = ['tomato' if l == lam else 'steelblue' for l in lam_vals]

fig, ax = plt.subplots(figsize=(10, 4))
bars = ax.bar(range(len(lam_vals)), energies, color=colors, edgecolor='k', lw=0.5)
ax.set_xticks(range(len(lam_vals)))
ax.set_xticklabels(lam_names, fontsize=9)
ax.set_ylabel("Photon energy (eV)", fontsize=12)
ax.set_yscale('log')
ax.set_title("Photon Energy Across the Spectrum
(this problem: mid-IR, highlighted red)", fontsize=11)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{\lambda = \frac{h}{p} = \frac{6.626 \times 10^{-34}}{5.00 \times 10^{-29}} \approx 1.33 \times 10^{-5}\text{ m} = 13.3\text{ µm (mid-IR)}}$$

$$\boxed{E = pc \approx 1.50 \times 10^{-20}\text{ J} \approx 0.094\text{ eV}}$$
