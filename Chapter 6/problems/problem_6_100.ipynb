# Problem 6.100 — Radius of Hydrogen Atom in the First Excited State

## Problem Statement

What is the radius of a hydrogen atom when the electron is in the **first excited state**?

---

## Background

In Bohr's model, the electron can only occupy specific circular orbits. The radius of the $n$-th orbit is:

$$r_n = n^2 a_0$$

where $a_0 = 5.292 \times 10^{-11}$ m is the **Bohr radius** — the radius of the ground state ($n = 1$) orbit.

Key facts:
- $n = 1$: ground state, $r_1 = a_0 = 0.529$ Å
- $n = 2$: **first excited state**, $r_2 = 4a_0$
- $n = 3$: second excited state, $r_3 = 9a_0$
- Orbital radius grows as $n^2$ — the atom expands dramatically in excited states

---

## Strategy

1. Identify the first excited state: $n = 2$
2. Apply $r_n = n^2 a_0 = 4a_0$
3. Convert to Ångströms for comparison


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── First excited state ────────────────────────────────────────────
n  = 2   # first excited state

# ── Bohr orbit radius ──────────────────────────────────────────────
r_n = n**2 * a0

print("=" * 50)
print(f"Bohr orbit radius:  r_n = n² · a₀")
print("=" * 50)
print(f"  n   = {n}  (first excited state)")
print(f"  a₀  = {a0:.4e} m  ({a0*1e10:.4f} Å)")
print(f"  r_n = n² · a₀")
print(f"  r_n = {n}² × {a0:.4e} m")
print(f"  r_n = {r_n:.4e} m")
print(f"  r_n = {r_n*1e10:.4f} Å")
print()
print(f"  Compared to ground state (n=1): r₁ = {a0*1e10:.3f} Å")
print(f"  Ratio r₂/r₁ = {r_n/(a0):.1f}  (4× larger — as expected from n²)")
```

## Results

| State | $n$ | Radius $r_n = n^2 a_0$ |
|-------|-----|------------------------|
| Ground | 1 | $a_0 = 0.529$ Å |
| **First excited** | **2** | **$4a_0 = 2.116$ Å** |
| Second excited | 3 | $9a_0 = 4.763$ Å |

---

## Physical Interpretation

- The first excited state atom is **4 times larger** than the ground state atom — even though we only went from $n=1$ to $n=2$.
- This rapid growth ($r \propto n^2$) means highly excited (Rydberg) atoms can reach macroscopic sizes: at $n=100$, the atom would be ~5300 Å = 530 nm across — comparable to a wavelength of visible light!
- In the ground state, the electron is tightly bound closest to the nucleus. Absorbing a photon promotes it to a larger, less-tightly-bound orbit.


```python
ns     = range(1, 8)
radii  = [n**2 * a0 * 1e10 for n in ns]   # in Ångströms

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left: bar chart of radii
ax = axes[0]
bars = ax.bar(list(ns), radii, color='steelblue', edgecolor='k', lw=0.5)
bars[1].set_color('tomato')   # highlight n=2
ax.set_xlabel("Quantum number n", fontsize=12)
ax.set_ylabel("Orbital radius (Å)", fontsize=12)
ax.set_title("Bohr Orbit Radii vs n
(n=2 first excited state highlighted in red)", fontsize=11)
for bar, r in zip(bars, radii):
    ax.text(bar.get_x() + bar.get_width()/2, r + 0.3, f"{r:.2f} Å", ha='center', fontsize=8)
ax.grid(axis='y', alpha=0.3)

# Right: r_n/a0 = n^2 curve
ax2 = axes[1]
n_cont = np.linspace(1, 7, 300)
ax2.plot(n_cont, n_cont**2, 'steelblue', lw=2.5, label=r'$r_n/a_0 = n^2$')
ax2.scatter(list(ns), [n**2 for n in ns], color='steelblue', s=60, zorder=5)
ax2.scatter([2], [4], color='tomato', s=120, zorder=6, label='n=2: r = 4a₀')
ax2.set_xlabel("Quantum number n", fontsize=12)
ax2.set_ylabel(r"$r_n / a_0$", fontsize=12)
ax2.set_title("Orbital Radius grows as $n^2$", fontsize=11)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.suptitle("Problem 6.100 — Hydrogen Atom Orbital Radii", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{r_2 = n^2 a_0 = (2)^2 \times 0.529\text{ Å} = 4 \times 0.529\text{ Å} = 2.116\text{ Å}}$$

The hydrogen atom in its first excited state has a radius of **2.116 Å** — exactly 4 times the Bohr radius.
