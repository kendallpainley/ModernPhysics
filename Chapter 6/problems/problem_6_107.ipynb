# Problem 6.107 — De Broglie Wavelength of an Electron at 5.0 × 10⁶ m/s

## Problem Statement

What is the de Broglie wavelength of an electron travelling at a speed of $v = 5.0 \times 10^6$ m/s?

---

## Background

In 1924, Louis de Broglie proposed that all matter has an associated wave, with wavelength:

$$\lambda = \frac{h}{p} = \frac{h}{mv}$$

For an electron moving at non-relativistic speeds ($v \ll c$), the momentum is $p = m_e v$, giving:

$$\lambda = \frac{h}{m_e v}$$

We first check whether relativistic corrections are needed by computing $\beta = v/c$. If $\beta \ll 0.1$, the non-relativistic formula is valid.

---

## Strategy

1. Compute $\beta = v/c$ to verify the non-relativistic approximation
2. Calculate momentum: $p = m_e v$
3. Calculate de Broglie wavelength: $\lambda = h/p$
4. Convert to convenient units (pm, Å) and identify the spectral region


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── Given value ────────────────────────────────────────────────────
v = 5.0e6    # speed [m/s]

# ── Check relativistic validity ────────────────────────────────────
beta = v / c

print("=" * 50)
print("Relativistic check")
print("=" * 50)
print(f"  v   = {v:.2e} m/s")
print(f"  c   = {c:.3e} m/s")
print(f"  β   = v/c = {beta:.4f}")
if beta < 0.1:
    print(f"  β < 0.1  →  non-relativistic approximation is valid ✓")
print()

# ── Momentum and wavelength ────────────────────────────────────────
p   = m_e * v
lam = h / p

print("=" * 50)
print("de Broglie wavelength:  λ = h / (m_e v)")
print("=" * 50)
print(f"  m_e = {m_e:.3e} kg")
print(f"  v   = {v:.2e} m/s")
print(f"  p   = m_e · v = {m_e:.3e} × {v:.2e} = {p:.4e} kg·m/s")
print(f"  λ   = h / p   = {h:.3e} / {p:.4e}")
print(f"  λ   = {lam:.4e} m")
print(f"  λ   = {lam*1e12:.4f} pm")
print(f"  λ   = {lam*1e10:.4f} Å")
print()
print(f"  This is in the X-ray / crystal lattice spacing range.")
print(f"  Electrons at this speed are used in electron diffraction experiments.")
```

## Results

| Quantity | Value |
|----------|-------|
| Speed $v$ | $5.0 \times 10^6$ m/s |
| $\beta = v/c$ | 0.0167 (non-relativistic) |
| Momentum $p = m_e v$ | $4.555 \times 10^{-24}$ kg·m/s |
| de Broglie wavelength $\lambda$ | **145.4 pm = 1.454 Å** |

---

## Physical Interpretation

- A wavelength of ~1.45 Å is comparable to atomic spacings in crystals (~1–5 Å), which is why such electrons produce clear **diffraction patterns** when scattered from crystal lattices (Davisson-Germer experiment).
- This is far smaller than visible light wavelengths, explaining why electron microscopes resolve atomic-scale features that light microscopes cannot.
- The speed of $5 \times 10^6$ m/s corresponds to electrons accelerated through about 71 V — a very accessible laboratory voltage.


```python
vs   = np.logspace(4, 8.5, 400)
lams = h / (m_e * vs) * 1e12   # pm

fig, ax = plt.subplots(figsize=(9, 5))
ax.loglog(vs, lams, 'steelblue', lw=2.5, label='λ = h/(m_e v)')
ax.axvline(v, color='tomato', ls='--', lw=2,
           label=f'v = 5×10⁶ m/s → λ = {lam*1e12:.1f} pm')
ax.scatter([v], [lam*1e12], color='tomato', s=100, zorder=5)

# Reference lines
ax.axhline(215, color='gray', ls=':', lw=1, alpha=0.7, label='Atom spacing ~2 Å')
ax.axhline(1000, color='purple', ls=':', lw=1, alpha=0.7, label='X-ray (1 nm)')

ax.set_xlabel("Electron speed (m/s)", fontsize=12)
ax.set_ylabel("de Broglie wavelength (pm)", fontsize=12)
ax.set_title("Electron de Broglie Wavelength vs Speed
"
             "Shorter wavelength = better resolving power", fontsize=12)
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{\lambda = \frac{h}{m_e v} = \frac{6.626 \times 10^{-34}}{(9.109 \times 10^{-31})(5.0 \times 10^6)} \approx 145\text{ pm} = 1.45\text{ Å}}$$

The de Broglie wavelength is **145 pm**, in the X-ray range and comparable to crystal lattice spacings — ideal for electron diffraction experiments.
