# Problem 6.89 — First Line of the Lyman Series

## Problem Statement

Calculate the wavelength of the first line in the Lyman series and show that this line lies in the ultraviolet part of the spectrum.

---

## Background

The **Rydberg formula** gives the wavelengths of all spectral emission lines of atomic hydrogen:

$$\frac{1}{\lambda} = R_H \left(\frac{1}{n_f^2} - \frac{1}{n_i^2}\right)$$

where:
- $R_H = 1.097 \times 10^7$ m⁻¹ is the Rydberg constant
- $n_f$ is the final (lower) energy level
- $n_i$ is the initial (higher) energy level, with $n_i > n_f$

The **Lyman series** corresponds to transitions ending at $n_f = 1$ (the ground state). These photons all carry enough energy to bridge the gap down to the lowest energy level — so they are the most energetic hydrogen lines and lie in the ultraviolet.

The **first line** of the Lyman series is the one with the longest wavelength (lowest energy) in the series, corresponding to the smallest transition: $n_i = 2 \to n_f = 1$.

---

## Strategy

1. Set $n_f = 1$, $n_i = 2$
2. Calculate $1/\lambda = R_H(1/1^2 - 1/2^2)$
3. Invert to find $\lambda$
4. Confirm $\lambda < 380$ nm (UV)


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── Quantum numbers ────────────────────────────────────────────────
nf = 1    # Lyman series — all end at n=1
ni = 2    # first line: smallest transition

# ── Rydberg formula ────────────────────────────────────────────────
inv_lam = R_H * (1/nf**2 - 1/ni**2)
lam     = 1 / inv_lam

print("=" * 50)
print("Rydberg formula:  1/λ = R_H (1/n_f² - 1/n_i²)")
print("=" * 50)
print(f"  n_f = {nf},  n_i = {ni}")
print(f"  1/n_f² - 1/n_i² = 1/{nf**2} - 1/{ni**2}")
print(f"                   = {1/nf**2:.4f} - {1/ni**2:.4f}")
print(f"                   = {1/nf**2 - 1/ni**2:.4f}")
print(f"  1/λ = {R_H:.4e} × {1/nf**2-1/ni**2:.4f}")
print(f"  1/λ = {inv_lam:.4e} m⁻¹")
print(f"  λ   = 1 / {inv_lam:.4e}")
print(f"  λ   = {lam:.4e} m = {lam*1e9:.2f} nm")
print()
if lam*1e9 < 380:
    print(f"  ✓  {lam*1e9:.2f} nm < 380 nm → ULTRAVIOLET confirmed")
```

## Results

| Quantity | Value |
|----------|-------|
| Transition | $n_i = 2 \to n_f = 1$ |
| $1/\lambda$ | $8.228 \times 10^6$ m⁻¹ |
| Wavelength $\lambda$ | **121.6 nm** |
| Spectral region | **Ultraviolet (UV)** |

---

## Physical Interpretation

- This line is called **Lyman-alpha** and is one of the most important lines in astrophysics. It is the dominant spectral feature of hydrogen in the universe.
- At 121.6 nm, it lies in the far UV — completely invisible to the naked eye.
- All Lyman series lines lie in the UV because they all involve transitions down to $n=1$, which require (or release) large amounts of energy.
- Each successive Lyman line ($n_i = 3, 4, 5, ...$) has a shorter wavelength and converges to the **Lyman limit** at 91.2 nm.


```python
# Show all Lyman series lines
nf_L   = 1
ni_vals = range(2, 9)
lams_L  = [1 / (R_H * (1/nf_L**2 - 1/ni**2)) * 1e9 for ni in ni_vals]
lam_limit = 1 / (R_H * 1/nf_L**2) * 1e9   # series limit

fig, ax = plt.subplots(figsize=(10, 3.5))

for lam_v, ni in zip(lams_L, ni_vals):
    lw    = 2.5 if ni == 2 else 1.5
    alpha = 1.0 if ni == 2 else 0.6
    col   = 'tomato' if ni == 2 else 'purple'
    ax.axvline(lam_v, color=col, lw=lw, alpha=alpha,
               label=f'n={ni}→1: {lam_v:.1f} nm' if ni <= 5 else None)

ax.axvline(lam_limit, color='black', ls=':', lw=1.2, label=f'Series limit: {lam_limit:.1f} nm')
ax.axvspan(380, 700, alpha=0.10, color='violet', label='Visible (380–700 nm)')
ax.axvspan(0, 380, alpha=0.05, color='indigo', label='UV')

ax.set_xlim(80, 450)
ax.set_ylim(0, 1)
ax.set_yticks([])
ax.set_xlabel("Wavelength (nm)", fontsize=12)
ax.set_title("Lyman Series — All lines lie in the Ultraviolet
(first line, n=2→1, highlighted in red)", fontsize=12)
ax.legend(fontsize=9, loc='upper right')
ax.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{\frac{1}{\lambda} = R_H\left(\frac{1}{1^2} - \frac{1}{2^2}\right) = R_H \cdot \frac{3}{4} \implies \lambda = \frac{4}{3R_H} \approx 121.6\text{ nm (UV)}}$$

The first Lyman line (Lyman-alpha) is at **121.6 nm**, firmly in the ultraviolet region of the spectrum.
