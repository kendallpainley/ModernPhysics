# Problem 6.67 — Work Function of Aluminum

## Problem Statement

Estimate the work function of aluminum, given that the wavelength of **304 nm** is the longest wavelength that a photon may have to eject a photoelectron from an aluminum photoelectrode.

---

## Background

The **photoelectric effect** requires photons to have enough energy to overcome the binding energy of electrons in a metal surface. This minimum energy is called the **work function** $\phi$.

At the **cutoff wavelength** $\lambda_c$, the photon energy exactly equals the work function — the ejected electron has zero kinetic energy:

$$hf_c = \phi \implies \phi = \frac{hc}{\lambda_c}$$

Any photon with $\lambda > \lambda_c$ (lower energy) cannot eject an electron. Any photon with $\lambda < \lambda_c$ ejects an electron with kinetic energy $K_{max} = hc/\lambda - \phi$.

---

## Strategy

1. The cutoff wavelength is given: $\lambda_c = 304$ nm
2. Apply $\phi = hc/\lambda_c$ using $hc = 1240$ eV·nm
3. Compare to the accepted value for aluminum (~4.08 eV)


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── Given value ────────────────────────────────────────────────────
lam_c = 304.0   # cutoff wavelength [nm]

# ── Work function ──────────────────────────────────────────────────
phi = hc / lam_c

print("=" * 50)
print("Work function:  ϕ = hc / λ_c")
print("=" * 50)
print(f"  hc    = {hc:.1f} eV·nm")
print(f"  λ_c   = {lam_c:.1f} nm")
print(f"  ϕ = hc / λ_c = {hc:.1f} / {lam_c:.1f}")
print(f"  ϕ = {phi:.4f} eV")
print()
print(f"  Accepted value for Al: ~4.08 eV")
print(f"  Our estimate:           {phi:.3f} eV")
print(f"  Percentage difference:  {abs(phi-4.08)/4.08*100:.1f}%")
```

## Results

| Quantity | Value |
|----------|-------|
| Cutoff wavelength $\lambda_c$ | 304 nm |
| Work function $\phi$ | **4.079 eV** |
| Accepted value (Al) | ~4.08 eV |

---

## Physical Interpretation

- The work function represents the minimum energy needed to free an electron from the metal surface.
- For aluminum, only UV photons (λ < 304 nm) can cause the photoelectric effect — visible light cannot.
- Metals with lower work functions (e.g. sodium, 2.46 eV) respond to visible light and are used in photocells and solar panels.
- Metals with higher work functions (e.g. platinum, 6.35 eV) require deep UV light.


```python
metals = {'Na': 2.46, 'Al': phi, 'Pb': 4.14, 'Zn': 4.31,
          'Fe': 4.50, 'Cu': 4.70, 'Ag': 4.73, 'Pt': 6.35}
cols = ['tomato' if m == 'Al' else 'steelblue' for m in metals]
lam_cuts = {m: 1240/v for m, v in metals.items()}

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ax = axes[0]
bars = ax.bar(list(metals.keys()), list(metals.values()), color=cols, edgecolor='k', lw=0.5)
ax.set_ylabel("Work function (eV)", fontsize=12)
ax.set_title("Work Functions of Common Metals
(Aluminum highlighted in red)", fontsize=11)
for bar, v in zip(bars, metals.values()):
    ax.text(bar.get_x() + bar.get_width()/2, v + 0.05, f"{v:.2f}", ha='center', fontsize=8)
ax.grid(axis='y', alpha=0.3)

ax2 = axes[1]
bars2 = ax2.bar(list(lam_cuts.keys()), list(lam_cuts.values()), color=cols, edgecolor='k', lw=0.5)
ax2.axhspan(380, 700, alpha=0.12, color='violet', label='Visible (380–700 nm)')
ax2.set_ylabel("Cutoff wavelength λ_c (nm)", fontsize=12)
ax2.set_title("Cutoff Wavelength for Each Metal", fontsize=11)
ax2.legend(fontsize=9)
ax2.grid(axis='y', alpha=0.3)

plt.suptitle("Problem 6.67 — Work Function of Aluminum", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{\phi = \frac{hc}{\lambda_c} = \frac{1240\text{ eV·nm}}{304\text{ nm}} \approx 4.08\text{ eV}}$$

The work function of aluminum is approximately **4.08 eV**, requiring UV photons (λ < 304 nm) to trigger the photoelectric effect.
