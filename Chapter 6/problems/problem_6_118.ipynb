# Problem 6.118 — Lorentz Factor and de Broglie Wavelength of a 50-GeV Electron

## Problem Statement

Find the Lorentz factor $\gamma$ and de Broglie wavelength for a **50-GeV** electron in a particle accelerator.

---

## Background

At particle accelerator energies, electrons are **highly relativistic** and we must use relativistic mechanics.

The total relativistic energy is:
$$E = K + E_0 = K + m_e c^2$$

The Lorentz factor is defined as:
$$\gamma = \frac{E}{E_0} = \frac{E}{m_e c^2}$$

The relativistic momentum is found from:
$$E^2 = (pc)^2 + (m_e c^2)^2 \implies p = \frac{\sqrt{E^2 - E_0^2}}{c}$$

The de Broglie wavelength is then:
$$\lambda = \frac{h}{p}$$

Since $K = 50$ GeV $\gg E_0 = 0.511$ MeV, the electron is ultra-relativistic and $p \approx E/c$.

---

## Strategy

1. Convert energies to consistent units (GeV)
2. Compute total energy $E = K + E_0$
3. Find $\gamma = E/E_0$
4. Find $p = \sqrt{E^2 - E_0^2}/c$
5. Find $\lambda = h/p$


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── Kinetic energy and rest energy ─────────────────────────────────
K_GeV  = 50.0                   # kinetic energy [GeV]
E0_GeV = 0.511e-3               # electron rest energy: 511 keV = 0.000511 GeV
E_GeV  = K_GeV + E0_GeV        # total energy [GeV]

print("=" * 55)
print("Energies")
print("=" * 55)
print(f"  K   = {K_GeV} GeV  (kinetic energy)")
print(f"  E₀  = {E0_GeV*1e3:.3f} MeV = {E0_GeV:.6f} GeV  (rest energy)")
print(f"  E   = K + E₀ = {E_GeV:.6f} GeV  ≈ {E_GeV:.4f} GeV")
print()

# ── Lorentz factor ─────────────────────────────────────────────────
gamma = E_GeV / E0_GeV

print("=" * 55)
print("Lorentz factor:  γ = E / E₀")
print("=" * 55)
print(f"  γ = {E_GeV:.4f} GeV / {E0_GeV:.6f} GeV")
print(f"  γ = {gamma:.1f}")
print()

# ── Relativistic momentum ──────────────────────────────────────────
p_GeV = np.sqrt(E_GeV**2 - E0_GeV**2)      # GeV/c
p_kg  = p_GeV * 1e9 * eV / c               # convert to SI [kg·m/s]

print("=" * 55)
print("Momentum:  p = √(E² - E₀²) / c")
print("=" * 55)
print(f"  p = √({E_GeV:.4f}² - {E0_GeV:.6f}²) GeV/c")
print(f"  p = {p_GeV:.6f} GeV/c  ≈  E/c  (ultra-relativistic)")
print(f"  p = {p_kg:.4e} kg·m/s")
print()

# ── de Broglie wavelength ──────────────────────────────────────────
lam = h / p_kg

print("=" * 55)
print("de Broglie wavelength:  λ = h/p")
print("=" * 55)
print(f"  λ = {h:.3e} / {p_kg:.4e}")
print(f"  λ = {lam:.4e} m")
print(f"  λ = {lam*1e18:.4f} am  (attometres)")
print(f"      (1 am = 10⁻¹⁸ m;  proton radius ≈ 0.85 fm = 850 am)")
```

## Results

| Quantity | Value |
|----------|-------|
| Kinetic energy $K$ | 50 GeV |
| Rest energy $E_0 = m_e c^2$ | 0.511 MeV |
| Total energy $E$ | ≈ 50.000511 GeV |
| Lorentz factor $\gamma$ | **97,847** |
| Momentum $p$ | ≈ $2.67 \times 10^{-17}$ kg·m/s |
| de Broglie wavelength $\lambda$ | **≈ 2.48 × 10⁻²⁰ m = 0.0248 am** |

---

## Physical Interpretation

- $\gamma \approx 98,000$ means this electron's total energy is ~98,000 times its rest energy — extremely relativistic.
- The wavelength is ~$2.5 \times 10^{-20}$ m — about 50,000 times smaller than a proton radius. Such electrons can probe the internal structure of protons and neutrons.
- At these energies, $K \gg E_0$, so $p \approx E/c$ and $\lambda \approx hc/E$ — the particle behaves similarly to a photon of the same energy.


```python
# γ and λ vs kinetic energy for electrons
K_range  = np.logspace(0, 5, 400)   # MeV
E0_MeV   = 0.511
E_range  = K_range + E0_MeV
gamma_r  = E_range / E0_MeV
p_range  = np.sqrt(E_range**2 - E0_MeV**2) * 1e6 * eV / c
lam_r    = h / p_range * 1e15   # femtometres

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ax = axes[0]
ax.loglog(K_range, gamma_r, 'steelblue', lw=2.5)
ax.axvline(K_GeV*1e3, color='tomato', ls='--', lw=2,
           label=f'K = 50 GeV → γ = {gamma:.0f}')
ax.scatter([K_GeV*1e3], [gamma], color='tomato', s=100, zorder=5)
ax.set_xlabel("Kinetic energy (MeV)", fontsize=12)
ax.set_ylabel("Lorentz factor γ", fontsize=12)
ax.set_title("Lorentz Factor vs Kinetic Energy (electron)", fontsize=11)
ax.legend(fontsize=10); ax.grid(True, alpha=0.3)

ax2 = axes[1]
ax2.loglog(K_range, lam_r, 'steelblue', lw=2.5)
ax2.axvline(K_GeV*1e3, color='tomato', ls='--', lw=2,
            label=f'K = 50 GeV → λ = {lam*1e18:.2f} am')
ax2.scatter([K_GeV*1e3], [lam*1e15], color='tomato', s=100, zorder=5)
ax2.axhline(0.85e3, color='gray', ls=':', lw=1, label='Proton radius (0.85 fm)')
ax2.set_xlabel("Kinetic energy (MeV)", fontsize=12)
ax2.set_ylabel("de Broglie wavelength (fm)", fontsize=12)
ax2.set_title("de Broglie Wavelength vs Kinetic Energy (electron)", fontsize=11)
ax2.legend(fontsize=9); ax2.grid(True, alpha=0.3)

plt.suptitle("Problem 6.118 — Relativistic Electron at 50 GeV", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{\gamma = \frac{E}{m_e c^2} = \frac{50.000\text{ GeV}}{0.511\text{ MeV}} \approx 97{,}847}$$

$$\boxed{\lambda = \frac{h}{p} \approx \frac{hc}{E} = \frac{1240\text{ eV·nm}}{50 \times 10^9\text{ eV}} \approx 2.48 \times 10^{-20}\text{ m}}$$

At 50 GeV, the electron is ultra-relativistic with $\gamma \approx 98,000$ and a de Broglie wavelength far smaller than any known particle — making it a powerful probe of sub-nuclear structure.
