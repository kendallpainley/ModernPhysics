# Problem 6.55 — Blackbody Radiation: Energy Quanta and Heating Time

## Problem Statement

**A 200-W heater emits 1.5-µm radiation.**

**(a)** What value of the energy quantum does it emit?  
**(b)** Assuming that the specific heat of a 4.0-kg body is 0.83 kcal/kg·K, how many of these photons must be absorbed by the body to increase its temperature by 2 K?  
**(c)** How long does the heating process in (b) take, assuming that all radiation emitted by the heater gets absorbed by the body?

---

## Background

Planck's hypothesis states that electromagnetic radiation is emitted in discrete packets called **quanta** (photons). The energy of each quantum depends only on the frequency (or equivalently, the wavelength) of the radiation:

$$E = hf = \frac{hc}{\lambda}$$

where $h = 6.626 \times 10^{-34}$ J·s is Planck's constant.

To heat a body, we need a total energy $Q = mc_s \Delta T$, where $c_s$ is the specific heat. Each absorbed photon contributes one quantum $E$, so the number of photons needed is $N = Q/E$.

Since the heater emits at power $P$, the time to deliver energy $Q$ is simply $t = Q/P$.

---

## Strategy

1. Convert wavelength to SI units: $\lambda = 1.5\,\mu\text{m} = 1.5 \times 10^{-6}$ m
2. **Part (a):** Compute $E = hc/\lambda$
3. **Part (b):** Compute total heat $Q = mc_s\Delta T$, then $N = Q/E$
4. **Part (c):** Compute $t = Q/P$


```python
import numpy as np
import matplotlib.pyplot as plt

# ── Physical constants ─────────────────────────────────────────────
h     = 6.626e-34    # Planck's constant        [J·s]
hbar  = h / (2 * np.pi)
c     = 2.998e8      # Speed of light           [m/s]
k_B   = 1.380e-23    # Boltzmann constant       [J/K]
sigma = 5.670e-8     # Stefan-Boltzmann const   [W/(m²·K⁴)]
Wien  = 2.898e-3     # Wien displacement const  [m·K]
eV    = 1.602e-19    # 1 eV in Joules
m_e   = 9.109e-31    # Electron mass            [kg]
m_p   = 1.673e-27    # Proton mass              [kg]
m_n   = 1.675e-27    # Neutron mass             [kg]
e_c   = 1.602e-19    # Elementary charge        [C]
hc    = 1240.0       # hc in eV·nm
a0    = 5.292e-11    # Bohr radius              [m]
R_H   = 1.097e7      # Rydberg constant         [1/m]
E0_H  = 13.6         # Hydrogen ionization energy [eV]
Lc    = 2.426e-12    # Compton wavelength       [m]
# ── Given values ───────────────────────────────────────────────────
lam  = 1.5e-6           # wavelength [m]
P    = 200.0            # heater power [W]
mass = 4.0              # body mass [kg]
c_s  = 0.83e3 * 4.184   # specific heat: 0.83 kcal/(kg·K) → J/(kg·K)
dT   = 2.0              # temperature rise [K]

# ── Part (a): Energy per quantum ───────────────────────────────────
E = h * c / lam

print("=" * 50)
print("Part (a): Energy per quantum")
print("=" * 50)
print(f"  E = hc / λ")
print(f"  E = ({h:.3e} J·s × {c:.3e} m/s) / {lam:.2e} m")
print(f"  E = {E:.4e} J")
print(f"  E = {E/eV:.4f} eV")
print()

# ── Part (b): Number of photons needed ─────────────────────────────
Q = mass * c_s * dT
N = Q / E

print("=" * 50)
print("Part (b): Number of photons needed")
print("=" * 50)
print(f"  Q = m · c_s · ΔT")
print(f"  Q = {mass} kg × {c_s:.2f} J/(kg·K) × {dT} K")
print(f"  Q = {Q:.2f} J")
print(f"  N = Q / E = {Q:.2f} / {E:.4e}")
print(f"  N = {N:.3e} photons")
print()

# ── Part (c): Heating time ─────────────────────────────────────────
t = Q / P

print("=" * 50)
print("Part (c): Heating time")
print("=" * 50)
print(f"  t = Q / P = {Q:.2f} J / {P} W")
print(f"  t = {t:.2f} s  ({t/60:.2f} min)")
```

## Results

| Part | Quantity | Value |
|------|----------|-------|
| (a) | Energy per quantum | $E = 1.324 \times 10^{-19}$ J = 0.827 eV |
| (b) | Photons needed | $N \approx 2.09 \times 10^{23}$ photons |
| (c) | Heating time | $t \approx 34.7$ s |

---

## Physical Interpretation

- Each 1.5-µm infrared photon carries only ~0.83 eV — far less than a visible-light photon (~2–3 eV). This is why infrared heaters emit many low-energy quanta.
- An enormous number of photons (~10²³) are needed to heat even a small body, confirming that individual quantum effects are completely undetectable in everyday heating.
- At 200 W, the process takes under a minute — consistent with everyday experience of how quickly a heater warms an object.


```python
# ── Visualization: Heating time vs heater power ────────────────────
powers = np.linspace(50, 1000, 400)
times  = Q / powers

fig, axes = plt.subplots(1, 2, figsize=(11, 5))

# Left: heating time vs power
ax = axes[0]
ax.plot(powers, times, 'steelblue', lw=2)
ax.axvline(P, color='tomato', ls='--', lw=1.5, label=f'P = {P:.0f} W → t = {t:.1f} s')
ax.scatter([P], [t], color='tomato', s=80, zorder=5)
ax.set_xlabel("Heater power (W)", fontsize=12)
ax.set_ylabel("Heating time (s)", fontsize=12)
ax.set_title(f"Time to raise body temperature by ΔT = {dT} K", fontsize=11)
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

# Right: photon energy vs wavelength context
lams_range = np.linspace(0.5e-6, 20e-6, 400)
E_range    = h * c / lams_range / eV

ax2 = axes[1]
ax2.plot(lams_range * 1e6, E_range, 'steelblue', lw=2)
ax2.axvline(lam * 1e6, color='tomato', ls='--', lw=1.5,
            label=f'λ = {lam*1e6:.1f} µm → E = {E/eV:.3f} eV')
ax2.scatter([lam * 1e6], [E / eV], color='tomato', s=80, zorder=5)
ax2.set_xlabel("Wavelength (µm)", fontsize=12)
ax2.set_ylabel("Photon energy (eV)", fontsize=12)
ax2.set_title("Infrared Photon Energy vs Wavelength", fontsize=11)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.suptitle("Problem 6.55 — Energy Quanta and Heating Time", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{E = 1.324 \times 10^{-19}\text{ J} \approx 0.827\text{ eV}}$$

$$\boxed{N \approx 2.09 \times 10^{23}\text{ photons}}$$

$$\boxed{t \approx 34.7\text{ s}}$$

Each 1.5-µm photon carries a tiny but fixed energy quantum. A macroscopic heating task requires an astronomical number of these quanta, delivered continuously at the rated power of the heater.
