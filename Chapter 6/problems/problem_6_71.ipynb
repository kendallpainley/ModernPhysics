# Problem 6.71 — Photoelectric Surface: Work Function, Cutoff Frequency, and Stopping Potential

## Problem Statement

A **600-nm** light falls on a photoelectric surface and electrons with the maximum kinetic energy of **0.17 eV** are emitted.

**(a)** Determine the work function of the surface.  
**(b)** Determine the cutoff frequency of the surface.  
**(c)** What is the stopping potential when the surface is illuminated with light of wavelength **400 nm**?

---

## Background

Einstein's photoelectric equation relates the photon energy, work function, and maximum kinetic energy of ejected electrons:

$$K_{max} = hf - \phi = \frac{hc}{\lambda} - \phi$$

- The **work function** $\phi$ is the minimum energy needed to free an electron from the surface
- The **cutoff frequency** $f_c$ is the minimum frequency for which $K_{max} = 0$: $f_c = \phi/h$
- The **stopping potential** $\Delta V_s$ is the reverse voltage needed to stop the fastest electrons: $K_{max} = e\Delta V_s$

---

## Strategy

1. **Part (a):** Rearrange Einstein's equation: $\phi = hc/\lambda - K_{max}$
2. **Part (b):** Use $f_c = \phi / h$
3. **Part (c):** Find $K_{max}$ at 400 nm using the work function, then $\Delta V_s = K_{max}/e$


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── Given values ───────────────────────────────────────────────────
lam1  = 600.0   # first wavelength [nm]
K1    = 0.17    # measured K_max at lam1 [eV]
lam2  = 400.0   # second wavelength for part (c) [nm]

# ── Part (a): Work function ────────────────────────────────────────
E_ph1 = hc / lam1
phi   = E_ph1 - K1

print("=" * 50)
print("Part (a): Work function")
print("=" * 50)
print(f"  E_photon = hc / λ = {hc:.1f} / {lam1:.1f} = {E_ph1:.4f} eV")
print(f"  ϕ = E_photon - K_max = {E_ph1:.4f} - {K1:.2f}")
print(f"  ϕ = {phi:.4f} eV")
print()

# ── Part (b): Cutoff frequency ─────────────────────────────────────
f_c = phi * eV / h

print("=" * 50)
print("Part (b): Cutoff frequency")
print("=" * 50)
print(f"  f_c = ϕ / h = ({phi:.4f} eV × {eV:.3e} J/eV) / {h:.3e} J·s")
print(f"  f_c = {f_c:.4e} Hz = {f_c/1e12:.3f} THz")
lam_c = hc / phi
print(f"  Corresponding cutoff wavelength: λ_c = {lam_c:.1f} nm")
print()

# ── Part (c): Stopping potential at 400 nm ─────────────────────────
E_ph2 = hc / lam2
K2    = E_ph2 - phi
Vs    = K2    # in volts (since K_max [eV] = e·ΔVs → ΔVs [V] = K_max [eV])

print("=" * 50)
print("Part (c): Stopping potential at λ = 400 nm")
print("=" * 50)
print(f"  E_photon = hc / λ = {hc:.1f} / {lam2:.1f} = {E_ph2:.4f} eV")
print(f"  K_max = E_photon - ϕ = {E_ph2:.4f} - {phi:.4f} = {K2:.4f} eV")
print(f"  ΔV_s = K_max / e = {Vs:.4f} V")
```

## Results

| Part | Quantity | Value |
|------|----------|-------|
| (a) | Work function $\phi$ | **1.897 eV** |
| (b) | Cutoff frequency $f_c$ | **4.59 × 10¹⁴ Hz = 459 THz** |
| (b) | Cutoff wavelength $\lambda_c$ | **654 nm** (red/near-IR) |
| (c) | Stopping potential at 400 nm | **1.203 V** |

---

## Physical Interpretation

- This surface has a low work function (~1.9 eV), so even red light (600 nm) can eject electrons — consistent with its cutoff wavelength of ~654 nm.
- At 400 nm (violet), the photons have significantly more energy, producing faster electrons that require a larger stopping potential (1.2 V) to halt.
- The stopping potential is independent of light intensity — it depends only on photon frequency.


```python
lams = np.linspace(hc/phi * 0.6, 750, 500)
K_arr = np.maximum(0, hc/lams - phi)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(lams, K_arr, 'steelblue', lw=2.5, label='K_max vs wavelength')
ax.axvline(600, color='tomato', ls='--', lw=2,
           label=f'600 nm → K_max = {K1:.2f} eV (given)')
ax.axvline(400, color='goldenrod', ls='--', lw=2,
           label=f'400 nm → V_s = {Vs:.3f} V (part c)')
ax.axvline(hc/phi, color='gray', ls=':', lw=1.5,
           label=f'Cutoff λ_c = {hc/phi:.1f} nm')
ax.axvspan(380, 700, alpha=0.08, color='violet', label='Visible range')
ax.set_xlabel("Wavelength (nm)", fontsize=12)
ax.set_ylabel("K_max (eV)  =  Stopping potential (V)", fontsize=12)
ax.set_title(f"Maximum Kinetic Energy vs Wavelength  (ϕ = {phi:.3f} eV)", fontsize=12)
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{\phi = \frac{hc}{\lambda} - K_{max} = \frac{1240}{600} - 0.17 \approx 1.90\text{ eV}}$$

$$\boxed{f_c = \frac{\phi}{h} \approx 4.59 \times 10^{14}\text{ Hz}}$$

$$\boxed{\Delta V_s = \frac{hc/\lambda_{400} - \phi}{e} \approx 1.20\text{ V}}$$

The work function determines both the cutoff frequency and the stopping potential at any given illumination wavelength.
