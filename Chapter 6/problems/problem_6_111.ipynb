# Problem 6.111 — Electron with Yellow-Light Properties

## Problem Statement

**(a)** What is the energy of an electron whose de Broglie wavelength is that of a photon of yellow light with wavelength **590 nm**?  
**(b)** What is the de Broglie wavelength of an electron whose energy is that of the photon of yellow light?

---

## Background

This problem highlights a crucial difference between photons and electrons:

**Photon** (massless): $E = hc/\lambda$ and $p = h/\lambda$, so $E = pc$

**Electron** (massive, non-relativistic): $p = h/\lambda$ and $K = p^2/(2m_e)$, so $K = h^2/(2m_e\lambda^2)$

At the same wavelength, a photon and an electron carry **very different energies** because their energy-momentum relations differ. Similarly, at the same energy, they have **very different wavelengths**.

---

## Strategy

**Part (a):** Given $\lambda_e = \lambda_{photon} = 590$ nm:
1. Find electron momentum: $p = h/\lambda$
2. Find electron kinetic energy: $K = p^2/(2m_e)$

**Part (b):** Given $K_e = E_{photon} = hc/\lambda$:
1. Find photon energy: $E_{ph} = hc/\lambda$
2. Find electron momentum from KE: $p = \sqrt{2m_e K}$
3. Find de Broglie wavelength: $\lambda_e = h/p$


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── Yellow light photon ────────────────────────────────────────────
lam_ph = 590e-9    # photon wavelength [m]
E_ph   = h * c / lam_ph  # photon energy [J]

print("=" * 55)
print("Reference: yellow light photon (λ = 590 nm)")
print("=" * 55)
print(f"  E_photon = hc/λ = {E_ph:.4e} J = {E_ph/eV:.4f} eV")
print()

# ── Part (a): Electron with same wavelength ────────────────────────
p_a = h / lam_ph
K_a = p_a**2 / (2 * m_e)

print("=" * 55)
print("Part (a): Electron with λ_e = 590 nm")
print("=" * 55)
print(f"  p   = h/λ = {h:.3e} / {lam_ph:.3e}")
print(f"  p   = {p_a:.4e} kg·m/s")
print(f"  K   = p² / (2m_e) = ({p_a:.4e})² / (2 × {m_e:.3e})")
print(f"  K   = {K_a:.4e} J  =  {K_a/eV:.6f} eV")
print()
print(f"  → This is {E_ph/K_a:.1f}× LESS than the photon energy at the same λ")
print()

# ── Part (b): Electron with same energy as photon ─────────────────
K_b   = E_ph
p_b   = np.sqrt(2 * m_e * K_b)
lam_b = h / p_b

print("=" * 55)
print("Part (b): Electron with K_e = E_photon = 2.11 eV")
print("=" * 55)
print(f"  K_e  = E_photon = {K_b:.4e} J = {K_b/eV:.4f} eV")
print(f"  p    = √(2 m_e K) = √(2 × {m_e:.3e} × {K_b:.4e})")
print(f"  p    = {p_b:.4e} kg·m/s")
print(f"  λ_e  = h/p = {h:.3e} / {p_b:.4e}")
print(f"  λ_e  = {lam_b:.4e} m  =  {lam_b*1e12:.4f} pm")
print()
print(f"  → Electron wavelength is {lam_ph/lam_b:.1f}× shorter than the photon wavelength")
```

## Results

| Comparison | Photon (590 nm) | Electron (part a) | Electron (part b) |
|---|---|---|---|
| Wavelength $\lambda$ | 590 nm | 590 nm (same) | **1.24 pm** (much shorter) |
| Energy | 2.11 eV | **$7.0 \times 10^{-6}$ eV** (much less) | 2.11 eV (same) |

---

## Physical Interpretation

- **Part (a):** At the same wavelength, the electron has *far less energy* than the photon. This is because $K_e = p^2/(2m_e)$ grows slowly with $p$, while $E_{ph} = pc$ grows much faster.
- **Part (b):** At the same energy, the electron has a *much shorter wavelength* than the photon. Because the electron is massive, it needs much higher momentum (and shorter $\lambda$) to carry the same energy.
- This comparison illustrates the fundamental difference between the energy-wavelength relations for massless photons and massive particles.


```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left: energy comparison
ax = axes[0]
kinds  = ['Photon\n(590 nm)', 'Electron (a)\nsame λ', 'Electron (b)\nsame E']
vals   = [E_ph/eV, K_a/eV, K_b/eV]
colors = ['gold', 'steelblue', 'tomato']
bars   = ax.bar(kinds, vals, color=colors, edgecolor='k', lw=0.5)
ax.set_yscale('log')
ax.set_ylabel("Energy (eV)", fontsize=12)
ax.set_title("Energy Comparison at 590-nm Yellow Light", fontsize=11)
for bar, v in zip(bars, vals):
    ax.text(bar.get_x() + bar.get_width()/2, v * 1.5, f"{v:.2e} eV",
            ha='center', fontsize=9)
ax.grid(axis='y', alpha=0.3)

# Right: wavelength comparison
ax2 = axes[1]
lams_compare = [lam_ph*1e9, lam_ph*1e9, lam_b*1e12 / 1000]  # all in nm
ax2.bar(kinds, [lam_ph*1e9, lam_ph*1e9, lam_b*1e9],
        color=colors, edgecolor='k', lw=0.5)
ax2.set_yscale('log')
ax2.set_ylabel("Wavelength (nm)", fontsize=12)
ax2.set_title("Wavelength Comparison", fontsize=11)
for i, (k, v) in enumerate(zip(kinds, [lam_ph*1e9, lam_ph*1e9, lam_b*1e9])):
    axes[1].text(i, v * 1.5, f"{v:.3f} nm", ha='center', fontsize=9)
ax2.grid(axis='y', alpha=0.3)

plt.suptitle("Problem 6.111 — Photon vs Electron: Same λ or Same E?", fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{\text{(a) Same } \lambda = 590\text{ nm}: \quad K_e = \frac{p^2}{2m_e} = \frac{h^2}{2m_e\lambda^2} \approx 7.0 \times 10^{-6}\text{ eV}}$$

$$\boxed{\text{(b) Same } E = 2.11\text{ eV}: \quad \lambda_e = \frac{h}{\sqrt{2m_e K}} \approx 1.24\text{ pm}}$$

A massive particle and a massless photon at the same wavelength carry very different energies — and at the same energy, they have very different wavelengths.
