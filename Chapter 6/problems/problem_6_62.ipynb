# Problem 6.62 — Photon Energies for Visible Light

## Problem Statement

The wavelengths of visible light range from approximately **400 to 750 nm**. What is the corresponding range of photon energies for visible light?

---

## Background

The energy of a photon depends on its frequency — and since frequency and wavelength are inversely related ($c = f\lambda$), energy and wavelength are also inversely related:

$$E = hf = \frac{hc}{\lambda}$$

A useful shortcut in electron-volts:

$$E\,(\text{eV}) = \frac{1240\text{ eV·nm}}{\lambda\,(\text{nm})}$$

Since $E \propto 1/\lambda$:
- **Shortest wavelength (violet, 400 nm) → highest photon energy**
- **Longest wavelength (red, 750 nm) → lowest photon energy**

---

## Strategy

1. Apply $E = hc/\lambda$ at both wavelength limits
2. Use $hc = 1240$ eV·nm for convenient calculation
3. State the energy range and identify the spectral region


```python
import numpy as np
import matplotlib.pyplot as plt

# ── Physical constants ─────────────────────────────────────────────
h     = 6.626e-34    # Planck's constant        [J·s]
hbar  = h / (2 * np.pi)
c     = 2.998e8      # Speed of light           [m/s]
k_B   = 1.380e-23    # Boltzmann constant       [J/K]
sigma = 5.670e-8     # Stefan-Boltzmann const   [W/(m²·K⁴)]
Wien  = 2.898e-3     # Wien displacement const  [m·K]
eV    = 1.602e-19    # 1 eV in Joules
m_e   = 9.109e-31    # Electron mass            [kg]
m_p   = 1.673e-27    # Proton mass              [kg]
m_n   = 1.675e-27    # Neutron mass             [kg]
e_c   = 1.602e-19    # Elementary charge        [C]
hc    = 1240.0       # hc in eV·nm
a0    = 5.292e-11    # Bohr radius              [m]
R_H   = 1.097e7      # Rydberg constant         [1/m]
E0_H  = 13.6         # Hydrogen ionization energy [eV]
Lc    = 2.426e-12    # Compton wavelength       [m]
# ── Wavelength limits ──────────────────────────────────────────────
lam_violet = 400e-9   # m  (violet end)
lam_red    = 750e-9   # m  (red end)

# ── Photon energies ────────────────────────────────────────────────
E_violet = h * c / lam_violet
E_red    = h * c / lam_red

print("=" * 50)
print("Photon energy:  E = hc / λ")
print("=" * 50)
print(f"  At λ = {lam_violet*1e9:.0f} nm (violet):")
print(f"    E = hc / λ = {hc:.0f} eV·nm / {lam_violet*1e9:.0f} nm")
print(f"    E = {E_violet/eV:.4f} eV  =  {E_violet:.4e} J")
print()
print(f"  At λ = {lam_red*1e9:.0f} nm (red):")
print(f"    E = hc / λ = {hc:.0f} eV·nm / {lam_red*1e9:.0f} nm")
print(f"    E = {E_red/eV:.4f} eV  =  {E_red:.4e} J")
print()
print(f"  Visible photon energy range: {E_red/eV:.3f} eV  to  {E_violet/eV:.3f} eV")
```

## Results

| Colour | Wavelength | Photon Energy |
|--------|-----------|---------------|
| Violet | 400 nm | **3.100 eV** |
| Red | 750 nm | **1.653 eV** |

Visible light photon energies range from approximately **1.65 eV to 3.10 eV**.

---

## Physical Interpretation

- All visible photons carry energies in the 1.6–3.1 eV range — enough to trigger photochemical reactions in the retina (vision) and in photosynthesis, but not enough to ionize most atoms.
- UV photons (> 3.1 eV) have enough energy to break chemical bonds in DNA, explaining sunburn.
- Infrared photons (< 1.65 eV) carry too little energy to trigger photochemical reactions — they only cause heating.


```python
# ── Visualization: rainbow plot of photon energy across visible ─────
def wl_to_rgb(wl_nm):
    """Approximate wavelength → RGB colour."""
    w = wl_nm
    if   380 <= w < 440: r, g, b = (440 - w) / 60, 0, 1
    elif 440 <= w < 490: r, g, b = 0, (w - 440) / 50, 1
    elif 490 <= w < 510: r, g, b = 0, 1, (510 - w) / 20
    elif 510 <= w < 580: r, g, b = (w - 510) / 70, 1, 0
    elif 580 <= w < 645: r, g, b = 1, (645 - w) / 65, 0
    elif 645 <= w <= 750: r, g, b = 1, 0, 0
    else: r, g, b = 0, 0, 0
    return (r, g, b)

lams   = np.linspace(380e-9, 750e-9, 500)
energies = h * c / lams / eV

fig, ax = plt.subplots(figsize=(10, 4))
for i in range(len(lams) - 1):
    col = wl_to_rgb(lams[i] * 1e9)
    ax.fill_between(lams[i:i+2] * 1e9, energies[i:i+2], alpha=0.85, color=col)

ax.set_xlabel("Wavelength (nm)", fontsize=12)
ax.set_ylabel("Photon energy (eV)", fontsize=12)
ax.set_title("Photon Energy across the Visible Spectrum
"
             "E = hc/λ  (violet = most energetic, red = least energetic)", fontsize=12)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{E_{violet} = \frac{1240\text{ eV·nm}}{400\text{ nm}} = 3.10\text{ eV}}$$

$$\boxed{E_{red} = \frac{1240\text{ eV·nm}}{750\text{ nm}} = 1.65\text{ eV}}$$

The visible spectrum spans photon energies from approximately **1.65 eV (red) to 3.10 eV (violet)**. Higher-frequency (shorter-wavelength) light carries more energetic photons.
