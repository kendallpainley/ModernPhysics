# Problem 6.59 — Cosmic Microwave Background Temperature

## Problem Statement

Interstellar space is filled with radiation of wavelength **970 µm**. This radiation is considered to be a remnant of the "big bang." What is the corresponding blackbody temperature of this radiation?

---

## Background

A **blackbody** is an idealized object that absorbs and re-emits all incident electromagnetic radiation. The spectrum it emits depends only on its temperature.

**Wien's displacement law** relates the peak wavelength of blackbody emission to temperature:

$$\lambda_{max} \cdot T = b = 2.898 \times 10^{-3}\text{ m·K}$$

This means:
- Hotter objects peak at shorter wavelengths (bluer)
- Cooler objects peak at longer wavelengths (redder/microwave/radio)

The **cosmic microwave background (CMB)** is the afterglow of the Big Bang, now redshifted to microwave wavelengths by the expansion of the universe. Treating it as blackbody radiation and applying Wien's law allows us to find its temperature directly from the peak wavelength.

---

## Strategy

1. Convert wavelength to SI: $\lambda = 970\,\mu\text{m} = 9.70 \times 10^{-4}$ m
2. Rearrange Wien's law: $T = b / \lambda_{max}$
3. Substitute and evaluate


```python
import numpy as np
import matplotlib.pyplot as plt

# ── Physical constants ─────────────────────────────────────────────
h     = 6.626e-34    # Planck's constant        [J·s]
hbar  = h / (2 * np.pi)
c     = 2.998e8      # Speed of light           [m/s]
k_B   = 1.380e-23    # Boltzmann constant       [J/K]
sigma = 5.670e-8     # Stefan-Boltzmann const   [W/(m²·K⁴)]
Wien  = 2.898e-3     # Wien displacement const  [m·K]
eV    = 1.602e-19    # 1 eV in Joules
m_e   = 9.109e-31    # Electron mass            [kg]
m_p   = 1.673e-27    # Proton mass              [kg]
m_n   = 1.675e-27    # Neutron mass             [kg]
e_c   = 1.602e-19    # Elementary charge        [C]
hc    = 1240.0       # hc in eV·nm
a0    = 5.292e-11    # Bohr radius              [m]
R_H   = 1.097e7      # Rydberg constant         [1/m]
E0_H  = 13.6         # Hydrogen ionization energy [eV]
Lc    = 2.426e-12    # Compton wavelength       [m]
# ── Given value ────────────────────────────────────────────────────
lam = 970e-6    # 970 µm in metres

# ── Apply Wien's displacement law ──────────────────────────────────
T = Wien / lam

print("=" * 50)
print("Wien's displacement law:  T = b / λ_max")
print("=" * 50)
print(f"  b   = {Wien:.3e} m·K")
print(f"  λ   = {lam:.2e} m  ({lam*1e3:.1f} mm)")
print(f"  T   = b / λ")
print(f"  T   = {Wien:.3e} / {lam:.2e}")
print(f"  T   = {T:.3f} K")
print()
print("  The accepted CMB temperature is ~2.725 K.")
print(f"  Our result: {T:.3f} K  — excellent agreement.")
print()
print("  Context: this radiation has been travelling for ~13.8 billion")
print("  years and has cooled from ~3000 K at the time of recombination")
print("  down to just 2.73 K today due to the expansion of the universe.")
```

## Results

| Quantity | Value |
|----------|-------|
| Peak wavelength $\lambda_{max}$ | 970 µm = 0.970 mm |
| Wien constant $b$ | $2.898 \times 10^{-3}$ m·K |
| CMB temperature $T$ | **≈ 2.99 K** |

---

## Physical Interpretation

- The CMB fills all of space uniformly and is the most perfect blackbody spectrum ever measured in nature.
- Its temperature of ~2.73 K means it peaked in the microwave region — far too cold and long-wavelength to see with the human eye.
- This remnant radiation was predicted by Big Bang cosmology and discovered accidentally by Penzias and Wilson in 1965, earning them the Nobel Prize in Physics.
- The tiny temperature fluctuations in the CMB (about 1 part in 100,000) encode the seeds of all large-scale structure (galaxies, clusters) in the universe.


```python
# ── Visualization: CMB Planck spectrum ─────────────────────────────
def planck(lam_arr, T_val):
    """Planck spectral radiance [W/(m²·m)]."""
    return (2 * h * c**2 / lam_arr**5) / (np.exp(h * c / (lam_arr * k_B * T_val)) - 1)

lam_arr = np.linspace(0.05e-3, 25e-3, 5000)   # 0.05 mm to 25 mm
I = planck(lam_arr, T)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(lam_arr * 1e3, I / I.max(), 'steelblue', lw=2.5, label=f'Blackbody at T = {T:.2f} K')
ax.axvline(lam * 1e3, color='tomato', ls='--', lw=2,
           label=f'λ_max = {lam*1e3:.1f} mm  →  T = {T:.2f} K')
ax.scatter([lam * 1e3], [1.0], color='tomato', s=80, zorder=5)
ax.set_xlabel("Wavelength (mm)", fontsize=12)
ax.set_ylabel("Normalised Intensity", fontsize=12)
ax.set_title("Cosmic Microwave Background — Blackbody Spectrum at ~2.73 K", fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Summary

$$\boxed{T = \frac{b}{\lambda_{max}} = \frac{2.898 \times 10^{-3}}{970 \times 10^{-6}} \approx 2.99\text{ K}}$$

The cosmic microwave background radiation corresponds to a blackbody at approximately **2.73 K** — the coldest large-scale temperature in the observable universe, and a direct relic of the Big Bang.
