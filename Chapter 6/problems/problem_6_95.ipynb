# Problem 6.95 — Emission Wavelengths from the Third Excited State

## Problem Statement

When a hydrogen atom is in its **third excited state**, what are the **shortest** and **longest** wavelengths of the photons it can emit?

---

## Background

Bohr's model identifies energy states by quantum number $n$:
- Ground state: $n = 1$  
- 1st excited state: $n = 2$
- 2nd excited state: $n = 3$
- **3rd excited state: $n = 4$** ← starting point here

When the atom emits a photon, the electron drops from $n = 4$ to any lower state ($n = 3, 2,$ or $1$). The photon energy equals the energy difference:

$$\Delta E = E_0\left(\frac{1}{n_f^2} - \frac{1}{n_i^2}\right) \quad \text{where } E_0 = 13.6\text{ eV}$$

The corresponding wavelength is $\lambda = hc/\Delta E$.

- **Largest $\Delta E$** (shortest $\lambda$): transition to $n_f = 1$ (largest energy gap)
- **Smallest $\Delta E$** (longest $\lambda$): transition to $n_f = 3$ (adjacent level)

---

## Strategy

1. Set $n_i = 4$, loop over all lower states $n_f = 1, 2, 3$
2. Compute $\Delta E = E_0(1/n_f^2 - 1/n_i^2)$ for each
3. Convert to wavelength: $\lambda = hc/\Delta E$ (in nm using $hc = 1240$ eV·nm)
4. Identify the shortest ($n_f = 1$) and longest ($n_f = 3$)


```python
import numpy as np
import matplotlib.pyplot as plt
h=6.626e-34; hbar=h/(2*np.pi); c=2.998e8; k_B=1.380e-23
sigma=5.670e-8; Wien=2.898e-3; eV=1.602e-19
m_e=9.109e-31; m_p=1.673e-27; m_n=1.675e-27; e_c=1.602e-19
hc=1240.0; a0=5.292e-11; R_H=1.097e7; E0_H=13.6; Lc=2.426e-12
# ── Initial state ──────────────────────────────────────────────────
ni = 4   # third excited state

print("=" * 55)
print(f"All emission transitions from n_i = {ni}")
print("=" * 55)

for nf in [3, 2, 1]:
    dE     = E0_H * (1/nf**2 - 1/ni**2)
    lam_nm = hc / dE
    region = "IR" if lam_nm > 700 else ("visible" if lam_nm >= 380 else "UV")
    tag    = ""
    if nf == 3: tag = "  ← longest λ (smallest ΔE)"
    if nf == 1: tag = "  ← shortest λ (largest ΔE)"
    print(f"  n={ni} → n={nf}:")
    print(f"    ΔE = {E0_H}({1/nf**2:.4f} - {1/ni**2:.4f}) = {dE:.4f} eV")
    print(f"    λ  = hc/ΔE = {hc}/{dE:.4f} = {lam_nm:.2f} nm  [{region}]{tag}")
    print()
```

## Results

| Transition | $\Delta E$ | Wavelength | Region |
|-----------|------------|-----------|--------|
| $n=4 \to n=3$ | 0.660 eV | **1875 nm** | Infrared (Paschen series) |
| $n=4 \to n=2$ | 2.550 eV | **486 nm** | Visible — blue-green (Balmer series) |
| $n=4 \to n=1$ | 12.75 eV | **97.3 nm** | Ultraviolet (Lyman series) |

- **Longest wavelength:** 1875 nm (infrared, Paschen-alpha line)
- **Shortest wavelength:** 97.3 nm (ultraviolet, Lyman series)

---

## Physical Interpretation

- From $n=4$, the atom can emit photons in three different series simultaneously (Lyman, Balmer, Paschen).
- The 486-nm blue-green line is the Balmer $\beta$ line — one of the four visible lines in the hydrogen spectrum.
- The 1875-nm infrared line (Paschen-alpha) is in the thermal infrared, detectable with IR sensors but invisible to the eye.


```python
ni = 4
nf_vals  = [1, 2, 3]
lams_nm  = [hc / (E0_H * (1/nf**2 - 1/ni**2)) for nf in nf_vals]
dE_vals  = [E0_H * (1/nf**2 - 1/ni**2) for nf in nf_vals]
colors   = ['purple', 'steelblue', 'tomato']
labels   = [f'n=4→{nf}: {l:.0f} nm' for nf, l in zip(nf_vals, lams_nm)]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: spectral line positions
ax = axes[0]
for lam_v, col, lab in zip(lams_nm, colors, labels):
    ax.axvline(lam_v, color=col, lw=3, label=lab)
ax.axvspan(380, 700, alpha=0.10, color='violet', label='Visible')
ax.set_xlim(50, 2100)
ax.set_ylim(0, 1)
ax.set_yticks([])
ax.set_xlabel("Wavelength (nm)", fontsize=12)
ax.set_title("Emission Wavelengths from n = 4", fontsize=12)
ax.legend(fontsize=10)

# Right: energy level diagram
ax2 = axes[1]
ax2.set_xlim(0, 4); ax2.set_ylim(-14.5, 0.5); ax2.axis('off')
levels = {1: -13.6, 2: -3.40, 3: -1.51, 4: -0.85}
for n, E in levels.items():
    ax2.axhline(E, xmin=0.15, xmax=0.85, color='k', lw=1.5)
    ax2.text(0.1, E, f'n={n}', ha='right', va='center', fontsize=11)
    ax2.text(3.9, E, f'{E:.2f} eV', ha='right', va='center', fontsize=9, color='gray')
for nf, col in zip([1, 2, 3], colors):
    lam_v = hc / (E0_H * (1/nf**2 - 1/4**2))
    ax2.annotate('', xy=(2.0, levels[nf]), xytext=(2.0, levels[4]),
                 arrowprops=dict(arrowstyle='->', color=col, lw=2.5))
    ax2.text(2.15, (levels[nf]+levels[4])/2, f'{lam_v:.0f} nm', fontsize=9, color=col)
ax2.set_title("Energy Level Diagram", fontsize=12)

plt.suptitle("Problem 6.95 — Emission from Third Excited State (n=4)", fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Summary

From the third excited state ($n = 4$):

$$\boxed{\lambda_{longest} = 1875\text{ nm (n=4→3, Paschen-alpha, infrared)}}$$

$$\boxed{\lambda_{shortest} = 97.3\text{ nm (n=4→1, Lyman series, UV)}}$$

The 486-nm visible line (n=4→2, Balmer-beta) is also emitted and appears blue-green in hydrogen's emission spectrum.
