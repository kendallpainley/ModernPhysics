# Introduction

Two of the most revolutionary concepts of the twentieth century were the description of light as a collection of particles, and the treatment of particles as waves. These wave properties of matter have led to the discovery of technologies such as electron microscopy, which allows us to examine submicroscopic objects such as grains of pollen.

# 6.1 | Blackbody Radiation

All bodies emit electromagnetic radiation over a range of wavelengths. When a body is heated and its temperature rises, the perceived wavelength of its emitted radiation changes from infrared to red, then orange, and so forth. This is the underlying principle of the incandescent light bulb: a hot metal filament glows red, and when heating continues, its glow eventually covers the entire visible portion of the electromagnetic spectrum.

The temperature $T$ of the emitting object determines the wavelength at which the radiated energy is at its maximum. The Sun, whose surface temperature is between 5000 K and 6000 K, radiates most strongly near 560 nm in the visible part of the spectrum. Your body, at about 300 K, radiates most strongly in the infrared.

#### Blackbody

At thermodynamic equilibrium, the rate at which an object absorbs radiation equals the rate at which it emits it. A perfect absorber absorbs all electromagnetic radiation incident on it — such an object is called a **blackbody**. A close realization is a small hole in the wall of a sealed cavity radiator. Electromagnetic waves emitted by a blackbody are called **blackbody radiation**.

#### Wien's Displacement Law

Two important laws summarize the experimental findings of blackbody radiation. The first is **Wien's displacement law** — the hotter the body, the shorter the wavelength of peak emission:

$$\lambda_{max} \cdot T = 2.898 \times 10^{-3} \text{ m·K}$$

where $\lambda_{max}$ is the peak wavelength and $T$ is temperature in kelvins. This law lets us estimate stellar temperatures from the color of their light. A blue star (short peak wavelength) is hotter than a red star.

#### Stefan's Law

The second experimental relation is **Stefan's law**, which gives the total power radiated across the entire spectrum:

$$P(T) = \sigma A T^4$$

where $A$ is the surface area of the blackbody, $T$ is its temperature in kelvins, and $\sigma$ is the **Stefan–Boltzmann constant**:

$$\sigma = 5.670 \times 10^{-8} \text{ W/(m}^2\text{·K}^4\text{)}$$

Total radiated power scales as the *fourth power* of temperature, so even modest increases in temperature produce dramatically larger emitted power.

#### The Ultraviolet Catastrophe

In a classical approach treating radiation as waves, the result is the **Rayleigh–Jeans law**. However, this law fails at short wavelengths — it predicts that radiation intensity increases without bound as wavelength decreases, a divergence called the **ultraviolet catastrophe**. This shows how classical physics fails to explain blackbody radiation.

#### Planck's Quantum Hypothesis

The blackbody radiation problem was solved in 1900 by Max Planck. His innovative idea was that atomic oscillators inside the cavity walls can have only *discrete* values of energy, and therefore the radiation can exchange energy with the walls only in discrete amounts.

Planck assumed that the energy of an oscillator $E_n$ can take only discrete, or quantized, values:

$$E_n = nhf, \quad \text{where } n = 1, 2, 3, \ldots$$

In this equation, $f$ is the frequency of Planck's oscillator, $n$ is called the **quantum number**, and $h$ is **Planck's constant**:

$$h = 6.626 \times 10^{-34} \text{ J·s} = 4.136 \times 10^{-15} \text{ eV·s}$$

Each discrete energy value corresponds to a **quantum state** of a Planck oscillator. Consecutive quantum states are separated by an energy jump:

$$\Delta E = hf$$

The absorption process sends the oscillator to a higher quantum state; the emission process sends it to a lower one. Whatever energy is exchanged must be an integer multiple of $hf$.

When included in the computation of the energy density of a blackbody, Planck's hypothesis gives the following theoretical expression for the power intensity of emitted radiation per unit wavelength — **Planck's blackbody radiation law**:

$$I(\lambda, T) = \frac{2\pi hc^2}{\lambda^5} \cdot \frac{1}{e^{hc/\lambda k_B T} - 1}$$

where $c$ is the speed of light in vacuum and $k_B = 1.380 \times 10^{-23}$ J/K is Boltzmann's constant.

# 6.2 | Photoelectric Effect

When a metal surface is exposed to a monochromatic electromagnetic wave of sufficiently short wavelength (equivalently, above a threshold frequency), the incident radiation is absorbed and the exposed surface emits electrons. This phenomenon is known as the **photoelectric effect**. Electrons emitted in this process are called **photoelectrons**.

#### Characteristics of the Photoelectric Effect

The photoelectric effect has three important characteristics that cannot be explained by classical physics:

1. **The absence of lag time** — Electrons are emitted almost instantaneously, even at very low intensities. Classical physics would predict a significant buildup time at low energy.

2. **Independence of kinetic energy on intensity** — The maximum kinetic energy of photoelectrons is independent of light intensity. Classical theory predicts higher intensity should produce higher kinetic energy.

3. **Presence of a cut-off frequency** — There is a minimum frequency below which no photocurrent occurs regardless of intensity. Classical theory predicts no such threshold.

The largest kinetic energy of photoelectrons can be directly measured through the **stopping potential** $\Delta V_s$:

$$K_{max} = e\Delta V_s$$

#### The Work Function and Einstein's Explanation

The photoelectric effect was explained in 1905 by Albert Einstein. He postulated that light itself consists of energy quanta — **photons**. A photon is a particle of light. Each photon moves at the speed of light and carries an energy quantum:

$$E_f = hf$$

In the photoelectric effect, each photon gives away *all* of its energy to only one electron — an "all or nothing" transfer, with no fractional transfers. The energy balance for an electron that receives energy $E_f$ from a photon is:

$$E_f = K_{max} + \phi$$

where $\phi$ is the **work function** of the metal — the energy needed to detach a photoelectron from the surface. Each metal has its own characteristic work function. Solving for $K_{max}$:

$$K_{max} = hf - \phi$$

This equation directly explains all three characteristics:
- Interactions are instantaneous because they occur one-on-one between individual photons and electrons.
- $K_{max}$ depends only on frequency $f$, not on intensity (which controls the number of photons, not their individual energy).
- Below the **cut-off frequency** $f_c$, no photon has enough energy to eject an electron:

$$f_c = \frac{\phi}{h}$$

The corresponding **cut-off wavelength** is:

$$\lambda_c = \frac{hc}{\phi}$$

where $hc = 1240$ eV·nm. When incident radiation has wavelengths longer than $\lambda_c$, the photoelectric effect does not occur.

# 6.3 | The Compton Effect

Beyond 1905, Einstein suggested that freely propagating electromagnetic waves consist of photons that are particles of light in the same sense that electrons or other massive particles are particles of matter.

#### Momentum of a Photon

A photon is massless. According to the theory of special relativity, any particle obeys the relativistic energy equation:

$$E^2 = p^2c^2 + m_0^2c^4$$

For a photon, we set $m_0 = 0$, which gives the momentum of a photon:

$$p_f = \frac{E_f}{c}$$

Since $E_f = hf = hc/\lambda$, the photon's momentum can be written as:

$$p_f = \frac{h}{\lambda}$$

Including direction, the photon's momentum is expressed as a vector:

$$\vec{p}_f = \hbar\vec{k}$$

where $\hbar = h/2\pi$ is the **reduced Planck's constant** and $\vec{k}$ is the **wave vector** with magnitude $k = 2\pi/\lambda$.

#### The Compton Effect

The **Compton effect** is the term used for an unusual result observed when X-rays are scattered on some materials. By classical theory, the scattered radiation should have the same wavelength as the incident radiation. Contrary to this, observations show that when X-rays scatter off materials such as graphite, the scattered X-rays have a longer wavelength. This phenomenon was studied experimentally by Arthur H. Compton and explained in 1923.

Compton treated the incident X-ray radiation as a stream of photons. An incoming photon collides with a loosely bound (effectively free) valence electron in the graphite target, transfers some of its energy and momentum to the electron, and emerges as a scattered photon with lower energy and hence longer wavelength.

#### Derivation of the Compton Shift

Compton derived the wavelength shift by assuming both the photon and electron are relativistic particles and that the collision conserves:
1. Linear momentum
2. Total relativistic energy

---

**Conservation of energy** (before and after collision):

$$E_f + m_0c^2 = \tilde{E}_f + E$$

where $E_f$ is the energy of the incident photon, $m_0c^2$ is the rest mass energy of the electron (initially at rest), $\tilde{E}_f$ is the energy of the scattered photon, and $E$ is the total energy of the recoiling electron.

---

**Conservation of linear momentum:**

$$\vec{p}_f = \tilde{\vec{p}}_f + \vec{p}$$

---

After algebraic manipulation of these two equations, the result for the **Compton shift** is:

$$\Delta\lambda = \lambda' - \lambda = \frac{h}{m_0 c}(1 - \cos\theta)$$

where $\theta$ is the scattering angle. The factor $h/m_0c$ is called the **Compton wavelength of the electron**:

$$\lambda_c = \frac{h}{m_0 c} = 0.00243 \text{ nm} = 2.43 \text{ pm}$$

Written in terms of $\lambda_c$:

$$\Delta\lambda = \lambda_c(1 - \cos\theta)$$

> The Compton shift $\Delta\lambda$ depends only on the scattering angle $\theta$, not on the wavelength of the incident radiation. The largest possible shift occurs at $\theta = 180°$ (backscattering):
>
> $$(\Delta\lambda)_{max} = 2\lambda_c = 4.86 \text{ pm}$$
>
> The nonshifted peak observed in experiments is due to photon collisions with tightly bound inner electrons, which effectively means the photon collides with the whole atom. The shift in that case is four orders of magnitude smaller and negligible.

The Compton effect has an important place in the history of physics because it shows that electromagnetic radiation cannot be explained as a purely wave phenomenon, and provided convincing evidence that photons behave like particles with momentum.


# 6.4 | Bohr's Model of the Hydrogen Atom

Bohr's model of the hydrogen atom, proposed in 1913, is the first model of atomic structure that correctly explained the radiation spectra of atomic hydrogen.

#### Atomic Spectra

When white light passes through a gas, some wavelengths are absorbed, producing an **absorption spectrum** — dark lines on a continuous background. When light is emitted by a gas, we observe an **emission spectrum** — colorful lines on a black background. Each chemical element has its own characteristic emission spectrum, and the positions of emission lines exactly match the positions of its absorption lines.

The simplest spectrum belongs to the hydrogen atom. An empirical formula discovered in 1885 by Johann Balmer describes the positions of the hydrogen emission lines in the **Balmer series**:

$$\frac{1}{\lambda} = R_H\left(\frac{1}{2^2} - \frac{1}{n^2}\right)$$

where $R_H = 1.09737 \times 10^7$ m$^{-1}$ is the **Rydberg constant** and $n = 3, 4, 5, 6$ for the four visible lines.

The more general **Rydberg formula** describes all hydrogen emission lines:

$$\frac{1}{\lambda} = R_H\left(\frac{1}{n_f^2} - \frac{1}{n_i^2}\right), \quad n_i = n_f + 1, n_f + 2, n_f + 3, \ldots$$

Different values of $n_f$ give different series: Lyman ($n_f = 1$), Balmer ($n_f = 2$), Paschen ($n_f = 3$), Brackett ($n_f = 4$), Pfund ($n_f = 5$), and Humphreys ($n_f = 6$).

#### Rutherford's Nuclear Model

In 1909, Rutherford, Marsden, and Geiger scattered $\alpha$-particles off a thin gold foil and found that a significant fraction were back-scattered toward the source. This result was only possible if most of the mass and the entire positive charge of the gold atom were concentrated in a tiny space inside the atom. In 1911, Rutherford proposed that an atom contains a positively charged **nucleus** of negligible size, with negative electrons located relatively far away.

#### Bohr's Three Postulates

Bohr's model combines classical mechanics with quantum concepts, built on three postulates:

1. The negative electron moves around the positive nucleus (proton) in a **circular orbit**. Not all classically possible orbits are available.

2. **First quantization condition:** In the $n$th orbit, the angular momentum $L_n$ of the electron can take only discrete values:

$$L_n = n\hbar, \quad n = 1, 2, 3, \ldots$$

Explicitly: $m_e v_n r_n = n\hbar$

3. **Second quantization condition:** Electron transitions between orbits involve photon absorption or emission:

$$hf = |E_n - E_m|$$

#### Electron Orbits

From Newtonian mechanics, the centripetal force on the electron equals the electrostatic attraction:

$$\frac{m_e v_n^2}{r_n} = \frac{1}{4\pi\varepsilon_0}\frac{e^2}{r_n^2}$$

Combining this with the first quantization condition gives the orbital radius:

$$r_n = a_0 n^2$$

where the **Bohr radius** $a_0$ is:

$$a_0 = \frac{4\pi\varepsilon_0\hbar^2}{m_e e^2} = 5.29 \times 10^{-11} \text{ m} = 0.529 \text{ Å}$$

The orbit size grows as the square of $n$: the second orbit is four times as large as the first, the third is nine times as large, and so on.

#### Electron Energies

The total energy of the electron in the $n$th orbit is:

$$E_n = -E_0\frac{1}{n^2}$$

where $E_0 = 13.6$ eV. The energy spectrum is quantized — it can only take on discrete values: $-E_0,\; -E_0/4,\; -E_0/9,\; -E_0/16, \ldots$

The smallest (most negative) energy is the **ground state energy**:

$$E_1 = -E_0 = -13.6 \text{ eV}$$

Higher energy states are **excited states**. The energy needed to remove the electron from the atom is the **ionization energy**:

$$E_\infty = -E_1 = E_0 = 13.6 \text{ eV}$$

#### Spectral Emission Lines of Hydrogen

When an electron makes a transition from the $n$th orbit to the $m$th orbit ($n > m$), the emitted photon carries away the energy difference. The wavelength of the emitted radiation is:

$$\frac{1}{\lambda} = \frac{E_0}{hc}\left(\frac{1}{m^2} - \frac{1}{n^2}\right)$$

The value $E_0/hc = 1.097 \times 10^7$ m$^{-1}$ is exactly the Rydberg constant $R_H$. In this way, Bohr's quantum model allows us to derive the Rydberg constant from first principles.

> Bohr's model also correctly predicts the spectra of **hydrogen-like ions** — atoms of element with atomic number $Z$ but with all electrons except one removed:
>
> $$r_n = \frac{a_0}{Z}n^2 \qquad E_n = -Z^2E_0\frac{1}{n^2}$$
>
> The limitation of early quantum theory is that it cannot describe atoms in which the number of orbiting electrons is larger than one.


# 6.5 | De Broglie's Matter Waves

Compton's formula established that an electromagnetic wave can behave like a particle of light when interacting with matter. In 1924, Louis de Broglie proposed that electrons and other particles of matter can behave like waves. Today, this idea is known as **de Broglie's hypothesis of matter waves**.

#### De Broglie Relations

According to de Broglie's hypothesis, any particle that has energy and momentum is a de Broglie wave of frequency $f$ and wavelength $\lambda$:

$$E = hf$$

$$\lambda = \frac{h}{p}$$

where $E$ and $p$ are the relativistic energy and momentum of a particle. These relations can also be expressed in terms of the wave vector $\vec{k}$ (with $k = 2\pi/\lambda$) and angular frequency $\omega = 2\pi f$:

$$E = \hbar\omega \qquad \vec{p} = \hbar\vec{k}$$

Wave theory tells us that a wave carries its energy with the **group velocity**. For matter waves, this group velocity is the velocity $u$ of the particle. The wave relation for matter waves is:

$$\lambda f = \frac{E}{p} = \frac{mc^2}{mu} = \frac{c^2}{u} = \frac{c}{\beta}$$

where $\beta = u/c$. When a particle is massless, $u = c$ and $\lambda f = c$ (the familiar light relation).

#### De Broglie Wavelengths

For a **nonrelativistic** particle (where $K \ll E_0 = m_0c^2$):

$$\lambda = \frac{h}{p} = \frac{h}{m_0 u}$$

For a **relativistic** particle with known kinetic energy $K$:

$$\lambda = \frac{hc}{\sqrt{K(K + 2E_0)}}$$

where $E_0 = m_0c^2$ is the rest mass energy.

> The de Broglie wavelengths of macroscopic objects are immeasurably small.

#### Justification of Bohr's Quantization Condition

Using the concept of the electron matter wave, de Broglie provided a physical rationale for Bohr's quantization of angular momentum. If an electron in the $n$th Bohr orbit behaves as a wave, the orbit circumference must fit an integer number of wavelengths — a standing-wave condition:

$$2\pi r_n = n\lambda$$

Using $\lambda = h/p$, this corresponds to linear momentum $p = nh/(2\pi r_n) = n\hbar/r_n$. In the circular orbit, the angular momentum must therefore be:

$$L_n = r_n p = r_n \cdot \frac{n\hbar}{r_n} = n\hbar$$

This is exactly Bohr's first quantization condition. Providing a physical explanation for this condition is a convincing theoretical argument for the existence of matter waves.

#### The Davisson–Germer Experiment

Experimental confirmation of matter waves came in 1927 when C. Davisson and L. Germer performed electron-scattering experiments. They scattered electrons off a crystalline nickel target and found clear **diffraction patterns** — the hallmark of wave behavior. For surface diffraction, the maximum intensity of the reflected electron beam occurs for angles $\varphi$ satisfying:

$$n\lambda = a\sin\varphi$$

where $a$ is the lattice spacing. With $a = 2.15$ Å and a first-order maximum at $\varphi \approx 50°$ for electrons accelerated through $\Delta V \approx 54$ V, the measured wavelength is $\lambda = 1.64$ Å. The de Broglie prediction gives $\lambda = 1.67$ Å — in excellent agreement.

Since the work of Davisson and Germer, de Broglie's hypothesis has been confirmed for numerous elementary particles, neutrons, and even large molecules such as $\text{C}_{60}$. The property of being a matter wave is not specific to charged particles but is true of all particles in motion.


# 6.6 | Wave-Particle Duality

The energy of radiation detected by a radio-signal receiving antenna comes as the energy of an electromagnetic wave. The same energy detected by a photocurrent in the photoelectric effect comes as individual photon particles. Is a photon a wave or a particle? Is an electron a wave or a particle? At our present state of knowledge, such questions do not have conclusive answers.

> **Wave-Particle Duality**
>
> Under some experimental conditions, a particle appears to act as a particle, and under different experimental conditions, a particle appears to act as a wave. Conversely, under some physical circumstances electromagnetic radiation acts as a wave, and under other physical circumstances, radiation acts as a beam of photons.

This dualistic interpretation is not new. Newton described light as corpuscles; Huygens described it as a wave. Young's 1803 double-slit experiment firmly established light as a wave. Maxwell's theory treats light as an electromagnetic wave. Yet Maxwell's classical view cannot explain blackbody radiation and the photoelectric effect, where light acts as a beam of photons.

#### Double-Slit Experiment with Electrons

The first double-slit experiment with a beam of electrons, performed by Claus Jönsson in Germany in 1961, demonstrated that a beam of electrons forms an interference pattern, which means electrons collectively behave as a wave. Subsequent experiments by Giulio Pozzi (1974) and Akira Tonomura (1989) with single electrons passing through slits one-by-one showed that interference fringes form gradually even when electrons pass individually. This demonstrates conclusively that electron-diffraction images are formed because of the wave nature of electrons.

#### Heisenberg's Uncertainty Principle

In quantum mechanics, de Broglie's idea of matter waves is taken as the fundamental property of all particles and given a statistical interpretation. A single particle is seen as a moving **wave packet**. We cannot measure its exact position in the same sense that we cannot pinpoint a location of a wave packet in a vibrating guitar string. The uncertainty $\Delta x$ in the particle's position is connected to the uncertainty $\Delta p$ in the simultaneous measurement of its linear momentum by **Heisenberg's uncertainty principle**:

$$\Delta x \Delta p \geq \frac{1}{2}\hbar$$

If we know the exact position of a body ($\Delta x = 0$), we cannot simultaneously know its momentum, because then $\Delta p \geq \hbar/(2\Delta x) \rightarrow \infty$. Heisenberg's principle sets the fundamental limit on the precision of simultaneous measurements of position and momentum — it is a law of nature, not a limitation of measurement technology.

> Heisenberg's uncertainty principle is a consequence of the wave nature of particles. At the quantum level, our perception is intrinsically limited.

#### Electron Microscopy

One practical application of wave-particle duality that exploits the wave properties of electrons is the **electron microscope**.

In the **transmission electron microscope (TEM)**, developed by Ernst Ruska in 1931, electrons are accelerated by a potential difference in an electron gun (giving up to 400 keV in kinetic energy), focused by electromagnetic lenses, and transmitted through a specimen. Modern high-resolution TEMs can have resolving power greater than 0.5 Å and magnifications higher than 50 million times. For comparison, the best resolving power obtained with light microscopy is about 97 nm.

The **scanning electron microscope (SEM)**, invented by Manfred von Ardenne in 1937, scatters a beam (up to 40 keV) off the surface of a sample. Surface topography is reconstructed by analyzing back-scattered electrons and emitted radiation. The resolving power of an SEM is better than 1 nm.

The wave-particle dual nature of matter and radiation is a declaration of our inability to describe physical reality within one unified classical theory. This led to the development of **quantum mechanics** by Bohr, Schrödinger, Heisenberg, and Dirac — which takes de Broglie's idea of matter waves as the fundamental property of all particles and gives it a statistical interpretation.
