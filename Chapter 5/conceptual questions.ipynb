 ### 5.1 Invariance of Physical Laws

**1. Which of Einstein's postulates of special relativity includes a concept that does not fit with the ideas of classical physics?**

The second postulate — that the speed of light is the same in all inertial frames — conflicts directly with classical physics. In Newtonian mechanics, velocities are simply additive (Galilean relativity), so an observer moving toward a light source would be expected to measure a higher speed of light than one moving away. The constancy of *c* regardless of the relative motion of the source and observer has no analog in classical physics and contradicts Galilean velocity addition entirely.

**2. Is Earth an inertial frame of reference? Is the sun? Justify your response.**

Neither Earth nor the sun is a perfect inertial frame. Earth rotates on its axis and orbits the sun, meaning it is constantly accelerating (centripetal acceleration). The sun also accelerates as it orbits the center of the Milky Way galaxy. However, for most practical purposes, both can be treated as approximately inertial frames when the accelerations involved are small compared to the effects being studied, since the accelerations are very small in magnitude compared to everyday forces.

**3. When you are flying in a commercial jet, it may appear to you that the airplane is stationary and Earth is moving beneath you. Is this point of view valid?**

Yes, this point of view is entirely valid. According to the first postulate of special relativity, the laws of physics are the same in all inertial frames of reference. Since a jet in level flight at constant speed is approximately an inertial frame, there is no physical experiment you could perform inside the jet that would determine whether the plane is moving and Earth is stationary, or vice versa. Both reference frames are equally valid for describing the motion.

### 5.3 Time Dilation

**4. (a) Does motion affect the rate of a clock as measured by an observer moving with it? (b) Does motion affect how an observer moving relative to a clock measures its rate?**

(a) No. An observer moving with the clock is in the same inertial reference frame as the clock. In that frame, the clock ticks at its normal rate. There is no way to distinguish "moving" from "at rest" within an inertial frame.

(b) Yes. An observer moving relative to the clock will measure the clock as ticking more slowly than normal. This is time dilation: the moving clock's tick rate appears slower to the observer in the different inertial frame.

**5. To whom does the elapsed time for a process seem to be longer, an observer moving relative to the process or an observer moving with the process? Which observer measures the interval of proper time?**

The elapsed time seems longer to the observer moving *relative* to the process. The observer who is stationary relative to the process (i.e., moving *with* the process) measures the proper time, which is the shortest possible time interval between the two events. All other observers in motion relative to the process measure a dilated (longer) time interval.

**6. (a) How could you travel far into the future of Earth without aging significantly? (b) Could this method also allow you to travel into the past?**

(a) You could travel at a speed very close to the speed of light. Due to time dilation, time in your frame would pass much more slowly than time on Earth. For example, if γ = 1000, one year in your frame would correspond to 1000 years on Earth. When you returned, Earth would have aged 1000 years while you had aged only one year, effectively sending you far into Earth's future.

(b) No, this method cannot allow travel into the past. Time dilation only slows time for the traveler relative to an outside observer; it never reverses it. The traveler still ages forward in their own frame and cannot return to an earlier time on Earth. Special relativity does not permit travel backward in time.

### 5.4 Length Contraction

**7. To whom does an object seem greater in length, an observer moving with the object or an observer moving relative to the object? Which observer measures the object's proper length?**

The object appears greater in length to the observer moving *with* the object (i.e., at rest relative to it). This observer measures the **proper length**, which is the longest possible measurement of the object's length. An observer moving relative to the object measures a contracted (shorter) length.

**8. Relativistic effects such as time dilation and length contraction are present for cars and airplanes. Why do these effects seem strange to us?**

Relativistic effects are present at any speed, but they become negligibly small at the speeds of everyday objects like cars and airplanes. For example, a car traveling at 100 km/h is moving at roughly 10⁻⁸ times the speed of light, making γ differ from 1 by an immeasurably tiny amount. Because humans have no direct sensory experience with speeds near *c*, these effects have never been part of our intuitive, everyday understanding of space and time. They seem strange because our intuitions were built entirely from experience in the classical (low-velocity) regime.

**9. Suppose an astronaut is moving relative to Earth at a significant fraction of the speed of light. (a) Does he observe the rate of his clocks to have slowed? (b) What change in the rate of earthbound clocks does he see? (c) Does his ship seem to him to shorten? (d) What about the distance between two stars that lie in the direction of his motion? (e) Do he and an earthbound observer agree on his velocity relative to Earth?**

(a) No. In his own rest frame, his clocks tick at their normal rate. He cannot detect any slowing of time within his own frame.

(b) From the astronaut's perspective, he is stationary and Earth is moving relative to him. Therefore, he observes earthbound clocks as running slowly (time dilation applies symmetrically during uniform relative motion).

(c) No. The ship does not appear shortened to him because he is at rest relative to it. Length contraction is only observed in the direction of motion for objects that are moving relative to the observer.

(d) Yes. The distance between the two stars in the direction of his motion would appear contracted to the astronaut. From his frame, those stars are moving relative to him, and the distance between them is length-contracted.

(e) Yes. Both observers agree on their relative speed. Relative velocity is a quantity that both frames must agree on, since it is symmetric — each measures the same magnitude of relative velocity, just in opposite directions.


### 5.7 Doppler Effect for Light

**10. Explain the meaning of the terms "red shift" and "blue shift" as they relate to the relativistic Doppler effect.**

A **red shift** occurs when the source of light is moving away from the observer. The observed wavelength is longer (shifted toward the red end of the visible spectrum) than the wavelength emitted by the source. A **blue shift** occurs when the source is moving toward the observer, producing a shorter observed wavelength (shifted toward the blue/violet end of the spectrum). These shifts occur because the relative motion between source and observer stretches or compresses the wave pattern of the light.

**11. What happens to the relativistic Doppler effect when relative velocity is zero? Is this the expected result?**

When *v* = 0, the formula $\lambda_\text{obs} = \lambda_s\sqrt{(1+v/c)/(1-v/c)}$ reduces to $\lambda_\text{obs} = \lambda_s$, meaning the observed wavelength equals the source wavelength — there is no Doppler shift. This is exactly the expected result: if there is no relative motion between the source and observer, the observer should detect the light at the same frequency and wavelength at which it was emitted.

**12. Is the relativistic Doppler effect consistent with the classical Doppler effect in the respect that λ_obs is larger for motion away?**

Yes. In both the classical and relativistic cases, motion of the source away from the observer produces a larger observed wavelength (red shift), and motion toward the observer produces a smaller observed wavelength (blue shift). The relativistic formula gives a different (and correct) magnitude than the classical formula, but the qualitative direction of the shift is consistent.

**13. All galaxies farther away than about 50 × 10⁶ ly exhibit a red shift in their emitted light that is proportional to distance, with those farther and farther away having progressively greater red shifts. What does this imply, assuming that the only source of red shift is relative motion?**

This implies that all distant galaxies are moving away from Earth, and that the farther away a galaxy is, the faster it is receding. This is Hubble's Law and is the primary evidence for the expansion of the universe. It implies that the universe is not static but is expanding in all directions from every point, which, when traced backward in time, leads to the conclusion that the universe originated from an extremely hot, dense state — the Big Bang.


### 5.8 Relativistic Momentum

**14. How does modern relativity modify the law of conservation of momentum?**

Modern relativity does not abandon conservation of momentum but redefines what momentum means. Classical momentum $\vec{p} = m\vec{u}$ is replaced by relativistic momentum $\vec{p} = \gamma m\vec{u}$, where $\gamma = 1/\sqrt{1 - u^2/c^2}$. With this redefinition, momentum is conserved in all inertial frames of reference. At low speeds, γ ≈ 1 and the relativistic formula reduces to the classical one.

**15. Is it possible for an external force to be acting on a system and relativistic momentum to be conserved? Explain.**

No, not in general. Conservation of relativistic momentum holds only when the net external force on the system is zero — exactly as in classical mechanics. If a net external force acts on the system, the total relativistic momentum of the system will change over time, just as in classical mechanics. The modification that relativity makes is to the definition of momentum itself, not to the fundamental relationship between force and the rate of change of momentum.


### 5.9 Relativistic Energy

**16. How are the classical laws of conservation of energy and conservation of mass modified by modern relativity?**

In modern relativity, mass and energy are no longer separately conserved; instead, mass-energy is conserved as a unified quantity through the relation *E* = *mc*². Mass can be converted into energy and vice versa. The classical law of conservation of mass is an approximation that holds when the energies involved are small compared to the rest energies of the particles. The classical law of conservation of energy is preserved in relativistic mechanics, but the total energy must include rest energy *mc*², kinetic energy, and all other forms, in addition to the possibility that mass can be converted to other forms of energy.

**17. What happens to the mass of water in a pot when it cools, assuming no molecules escape or are added? Is this observable in practice? Explain.**

When the water cools, it loses thermal energy. According to *E* = *mc*², this decrease in energy corresponds to a tiny decrease in rest mass: Δ*m* = Δ*E*/*c*². However, the change is immeasurably small. For example, cooling a kilogram of water by 100°C releases about 4.2 × 10⁵ J of energy, corresponding to a mass decrease of only about 4.7 × 10⁻¹² kg — far below any measurable precision in everyday instruments. So while the effect is real in principle, it is completely unobservable in practice.

**18. Consider a thought experiment. You place an expanded balloon of air on weighing scales outside in the early morning. The balloon stays on the scales, and you are able to measure changes in its mass. Does the mass of the balloon change as the day progresses? Discuss the difficulties in carrying out this experiment.**

Yes, in principle the mass of the balloon would change slightly. As the day progresses and the air warms, the air molecules gain thermal (kinetic) energy, and by *E* = *mc*², this represents a tiny increase in rest mass. Conversely, the actual air density inside the balloon also changes with temperature, affecting buoyancy. The practical difficulties are enormous: (1) the relativistic mass change would be many orders of magnitude smaller than any scale could possibly measure; (2) changes in buoyancy as air temperature changes would be far larger effects and would need to be carefully accounted for; (3) wind, vibration, and other environmental factors would swamp the tiny signal. The experiment is conceptually valid but completely impractical with any foreseeable technology.

**19. The mass of the fuel in a nuclear reactor decreases by an observable amount as it puts out energy. Is the same true for the coal and oxygen combined in a conventional power plant? If so, is this observable in practice for the coal and oxygen? Explain.**

Yes, the same principle applies to coal combustion — the products of combustion (CO₂, H₂O, and heat) have slightly less total mass than the coal and oxygen that were consumed. However, the fractional mass change in chemical reactions is roughly 10⁻¹⁰ (about one part in ten billion), compared to roughly 10⁻³ for nuclear reactions. The mass change in coal combustion is far too small to detect directly by weighing the reactants and products with any practical instrument. In a nuclear reactor, the percentage is large enough that it can be measured by monitoring fuel mass carefully over time.

**20. We know that the velocity of an object with mass has an upper limit of c. Is there an upper limit on its momentum? Its energy? Explain.**

There is no upper limit on the momentum or total energy of a massive object. As *u* → *c*, both γ → ∞ and therefore *p* = γ*mu* → ∞ and *E* = γ*mc*² → ∞. So while the velocity is bounded by *c*, ever-increasing amounts of energy and momentum can be added to the object without the velocity ever quite reaching *c*. The energy and momentum grow without bound as the velocity asymptotically approaches *c*.

**21. Given the fact that light travels at c, can it have mass? Explain.**

No, light (photons) cannot have mass. If a photon had mass *m* > 0, then by $E = \gamma mc^2$, reaching a velocity of exactly *c* would require γ → ∞ and hence infinite energy, which is impossible. Furthermore, a massive particle always travels at less than *c* in any reference frame. The energy and momentum of a photon are related by *E* = *pc*, consistent with the energy-momentum relation $E^2 = (pc)^2 + (mc^2)^2$ only if *m* = 0. Light is massless, and its constant speed *c* in all inertial frames is a consequence of this, not a contradiction.

**22. If you use an Earth-based telescope to project a laser beam onto the moon, you can move the spot across the moon's surface at a velocity greater than the speed of light. Does this violate modern relativity? (Note that light is being sent from the Earth to the moon, not across the surface of the moon.)**

No, this does not violate special relativity. The spot on the moon is not a physical object or a carrier of information — it is merely a geometric projection. No mass, energy, or information actually travels across the lunar surface at superluminal speed. Each photon still travels from Earth to the moon at speed *c*. The apparent motion of the spot is a geometric effect arising from changing the direction of the beam; it is analogous to the way a shadow can appear to move faster than *c* under certain conditions. Special relativity prohibits the transfer of mass, energy, and information faster than *c*, none of which occurs here.
