# Problem 5.23 — Time Dilation: Calculating the Lorentz factor γ

## Problem Statement

**(a)** What is γ if $v = 0.250c$?  
**(b)** What is γ if $v = 0.500c$?

---

## Background

The **Lorentz factor** (also called the relativistic factor) γ is the central quantity in special relativity. It appears in the time-dilation formula, the length-contraction formula, the relativistic momentum, and the total relativistic energy.

It is defined as:

$$\gamma = \frac{1}{\sqrt{1 - \dfrac{v^2}{c^2}}}$$

where:
- $v$ is the speed of the moving object (or frame)
- $c = 3.00 \times 10^8$ m/s is the speed of light

**Key observations:**
- When $v = 0$: $\gamma = 1$
- As $v \to c$: $\gamma \to \infty$
- γ is always $\geq 1$
- Relativistic effects become noticeable roughly when $v > 0.1c$

---

## Strategy

1. Identify the known quantity: $v$ expressed as a fraction of $c$
2. Compute $\beta = v/c$
3. Compute $\beta^2 = (v/c)^2$
4. Compute $1 - \beta^2$
5. Take the square root, then take the reciprocal

```python
import math

# Speed of light (m/s)
c = 3.00e8

# ── Part (a): v = 0.250c ──────────────────────────────────────────────
v_a = 0.250 * c          # velocity in m/s
beta_a = v_a / c         # dimensionless ratio v/c

gamma_a = 1 / math.sqrt(1 - beta_a**2)

print("=" * 50)
print("Part (a): v = 0.250c")
print("=" * 50)
print(f"  beta  = v/c       = {beta_a:.4f}")
print(f"  beta² = (v/c)²    = {beta_a**2:.6f}")
print(f"  1 - beta²         = {1 - beta_a**2:.6f}")
print(f"  sqrt(1 - beta²)   = {math.sqrt(1 - beta_a**2):.6f}")
print(f"  γ = 1/sqrt(...)   = {gamma_a:.4f}")
print()

# ── Part (b): v = 0.500c ──────────────────────────────────────────────
v_b = 0.500 * c
beta_b = v_b / c

gamma_b = 1 / math.sqrt(1 - beta_b**2)

print("=" * 50)
print("Part (b): v = 0.500c")
print("=" * 50)
print(f"  beta  = v/c       = {beta_b:.4f}")
print(f"  beta² = (v/c)²    = {beta_b**2:.6f}")
print(f"  1 - beta²         = {1 - beta_b**2:.6f}")
print(f"  sqrt(1 - beta²)   = {math.sqrt(1 - beta_b**2):.6f}")
print(f"  γ = 1/sqrt(...)   = {gamma_b:.4f}")
```

## Results

| Speed  | β = v/c | γ      |
|--------|---------|--------|
| 0.250c | 0.250   | 1.0328 |
| 0.500c | 0.500   | 1.1547 |

---

## Physical Interpretation

- At $v = 0.250c$, γ ≈ 1.033. This means a clock on a spaceship moving at 25% the speed of light would tick only about 3.3% slower than a stationary clock. Noticeable but modest.
- At $v = 0.500c$, γ ≈ 1.155. At half the speed of light, the clock runs about 15.5% slower.
- γ grows slowly at first, then accelerates dramatically as $v \to c$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot γ vs v/c to visualise how it grows
beta_vals = np.linspace(0, 0.9999, 5000)
gamma_vals = 1 / np.sqrt(1 - beta_vals**2)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(beta_vals, gamma_vals, 'steelblue', linewidth=2)

# Mark our two specific values
for beta, gamma, label in [(0.25, gamma_a, r'$v=0.250c$'), (0.50, gamma_b, r'$v=0.500c$')]:
    ax.plot(beta, gamma, 'ro', markersize=8)
    ax.annotate(f'{label}\nγ = {gamma:.4f}',
                xy=(beta, gamma), xytext=(beta + 0.04, gamma + 0.5),
                fontsize=10, color='darkred',
                arrowprops=dict(arrowstyle='->', color='darkred'))

ax.set_xlabel(r'$\beta = v/c$', fontsize=13)
ax.set_ylabel(r'Lorentz factor $\gamma$', fontsize=13)
ax.set_title(r'Lorentz Factor $\gamma$ vs Speed $v/c$', fontsize=14)
ax.set_xlim(0, 1.0)
ax.set_ylim(0.9, 10)
ax.axhline(1, color='gray', linestyle='--', linewidth=0.8, label='Classical limit (γ = 1)')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print("Plot saved.")
```

## Summary

$$\boxed{\gamma(0.250c) \approx 1.033} \qquad \boxed{\gamma(0.500c) \approx 1.155}$$

The plot above shows that γ stays close to 1 for low speeds (left side) and shoots toward infinity as $v \to c$ (right side). The two data points for this problem lie in the region where relativistic effects are present but not yet dramatic.
