# Problem 5.31 — Length Contraction: Spaceship Passing Earth

## Problem Statement

A spaceship is **200 m long as seen on board**. It moves by the Earth at $v = 0.970c$. **What is its length as measured by an earthbound observer?**

---

## Background

**Length contraction** is the decrease in the measured length of an object when it is observed from a reference frame in which it is moving. The object is shortened along its direction of motion.

The formula is:

$$L = L_0 \sqrt{1 - \frac{v^2}{c^2}} = \frac{L_0}{\gamma}$$

where:
- $L_0$ = **proper length** (measured at rest relative to the object)
- $L$ = contracted length seen by the observer for whom the object is moving
- $v$ = relative speed between object and observer

---

## Given

| Quantity | Value |
|----------|-------|
| Proper length (on board) | $L_0 = 200$ m |
| Speed relative to Earth | $v = 0.970c$ |
| Speed of light | $c = 3.00 \times 10^8$ m/s |

## Step-by-Step Solution

### 1. Identify proper length

The 200 m measurement is made **on board** the ship, so the crew is at rest relative to the ship. This is the **proper length** $L_0 = 200$ m.

### 2. Compute γ

$$\beta = \frac{v}{c} = 0.970$$

$$\gamma = \frac{1}{\sqrt{1 - (0.970)^2}} = \frac{1}{\sqrt{1 - 0.9409}} = \frac{1}{\sqrt{0.0591}} \approx 4.108$$

### 3. Apply length contraction

$$L = \frac{L_0}{\gamma} = \frac{200\text{ m}}{4.108} \approx 48.7\text{ m}$$

$$\boxed{L \approx 48.7\text{ m}}$$

---

## Physical Interpretation

The earthbound observer measures the spaceship as only about **24% of its rest length**. This is an enormous relativistic effect because $v = 0.970c$ is highly relativistic (γ ≈ 4.1).

Note that the crew members inside the ship measure the ship normally at 200 m — they are at rest relative to it. They would, however, observe the distance between Earth landmarks in the direction of travel to be length-contracted by the same factor.

## Summary

$$\boxed{L = L_0\sqrt{1 - (0.970)^2} = 200 \times 0.2432 \approx 48.7\text{ m}}$$

An earthbound observer measures the **200 m spaceship as only ~48.7 m long** because it is traveling at 97% of the speed of light. The crew on board still measure their ship as 200 m — length contraction is always *relative* to the observer's frame.
