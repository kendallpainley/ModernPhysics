# Problem 5.46 — Relativistic Velocity Addition: Canister Between Two Ships

## Problem Statement

If two spaceships are heading directly toward each other at $v = 0.800c$, **at what speed must a canister be shot from the first ship to approach the other ship at $0.999c$ as seen by the second ship?**

---

## Background

In classical mechanics, velocities simply add: if ship 1 moves at $+0.800c$ and fires a canister at $u'$ relative to itself, an observer on ship 2 would see the canister at $u = 0.800c + u'$. But this can exceed $c$, which is forbidden.

The **relativistic velocity addition** formula corrects this:

$$u = \frac{u' + v}{1 + \dfrac{v\,u'}{c^2}}$$

where:
- $v$ = velocity of frame S′ relative to frame S (speed of ship 1 as seen by ship 2)
- $u'$ = velocity of canister as measured in frame S′ (ship 1's frame)
- $u$ = velocity of canister as measured in frame S (ship 2's frame)

Here we need to **invert** this — we know $u$ (what ship 2 sees) and $v$ (relative ship speed), and we want $u'$ (canister speed relative to ship 1).

### Inverting the formula

Starting from $u = \dfrac{u' + v}{1 + v u'/c^2}$, solve for $u'$:

$$u\left(1 + \frac{v u'}{c^2}\right) = u' + v$$

$$u + \frac{v u u'}{c^2} = u' + v$$

$$u - v = u'\left(1 - \frac{v u}{c^2}\right)$$

$$\boxed{u' = \frac{u - v}{1 - \dfrac{v u}{c^2}}}$$

---

## Setup

Let ship 2 be the reference frame S. Ship 1 moves toward ship 2 at $v = 0.800c$.

Ship 2 observes the canister approaching at $u = 0.999c$.

We want: $u'$ = speed of canister as measured by ship 1.

## Step-by-Step Solution

### 1. Set up the reference frames

Let **ship 2 be the lab frame S**. In this frame:
- Ship 1 approaches at $v = +0.800c$
- The canister approaches at $u = +0.999c$ (toward ship 2)

We want to find $u'$, the canister's speed in **ship 1's frame S′**.

### 2. Apply the inverse velocity addition formula

$$u' = \frac{u - v}{1 - \dfrac{v u}{c^2}} = \frac{0.999c - 0.800c}{1 - (0.800)(0.999)} = \frac{0.199c}{1 - 0.7992} = \frac{0.199c}{0.2008}$$

$$\boxed{u' \approx 0.991c}$$

### 3. Note what classical mechanics would predict

Classically: $u' = u - v = 0.999c - 0.800c = 0.199c$ — but this is the speed in ship 2's frame minus ship 1's speed. Since the ships are approaching each other, the classical answer for what ship 1 fires is also just $0.199c$... but this does not transform back consistently. The relativistic result $0.991c$ is what's actually needed.

---

## Why This Is Interesting

Notice that $u' = 0.991c$ is very close to $c$. Even though ship 2 sees the canister at $0.999c$ and ship 1 is already rushing toward ship 2 at $0.800c$, the canister's speed relative to ship 1 is "only" $0.991c$, not something absurd like $0.199c$ or greater than $c$. Relativistic velocity addition always keeps all measured speeds below $c$.

## Summary

$$\boxed{u' = \frac{0.999c - 0.800c}{1 - (0.800)(0.999)} = \frac{0.199c}{0.2008} \approx 0.991c}$$

The canister must be fired at approximately **0.991c relative to ship 1** for ship 2 to observe it approaching at $0.999c$.
