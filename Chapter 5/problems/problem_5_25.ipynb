# Problem 5.25 — Time Dilation: Lifetime of π-Mesons in the Laboratory

## Problem Statement

Particles called **π-mesons** (pions) are produced by accelerator beams. If these particles travel at $v = 2.70 \times 10^8$ m/s and live $\Delta\tau = 2.60 \times 10^{-8}$ s when at rest relative to an observer, **how long do they live as viewed in the laboratory?**

---

## Background

This is an application of the **time dilation** formula. The pion's decay process measures **proper time** Δτ because the start and end of its life occur at the same location in the pion's rest frame.

A laboratory observer sees the pion moving, so the lab-frame time interval Δt is longer — the pion appears to live longer:

$$\Delta t = \gamma \, \Delta\tau = \frac{\Delta\tau}{\sqrt{1 - v^2/c^2}}$$

- Δτ = **proper time** (rest frame of pion)
- Δt = **dilated time** (lab frame)

---

## Given

| Quantity        | Value                                |
|-----------------|--------------------------------------|
| Speed of pion   | $v = 2.70 \times 10^8$ m/s           |
| Proper lifetime | $\Delta\tau = 2.60 \times 10^{-8}$ s |
| Speed of light  | $c = 3.00 \times 10^8$ m/s           |

## Step-by-Step Solution

### 1. Compute the Lorentz factor

$$\beta = \frac{v}{c} = \frac{2.70 \times 10^8}{3.00 \times 10^8} = 0.900$$

$$\gamma = \frac{1}{\sqrt{1 - \beta^2}} = \frac{1}{\sqrt{1 - (0.900)^2}} = \frac{1}{\sqrt{1 - 0.810}} = \frac{1}{\sqrt{0.190}} \approx 2.294$$

### 2. Apply time dilation

$$\Delta t = \gamma \,\Delta\tau = 2.294 \times 2.60 \times 10^{-8}\text{ s}$$

$$\boxed{\Delta t \approx 5.96 \times 10^{-8}\text{ s}}$$

---

## Physical Significance

Without time dilation, a pion traveling at $v = 0.900c$ with a proper lifetime of $2.60 \times 10^{-8}$ s would travel at most:

$$d_{\text{no dilation}} = v \cdot \Delta\tau = (0.900)(3.00 \times 10^8)(2.60 \times 10^{-8}) \approx 7.02\text{ m}$$

With time dilation, it travels:

$$d_{\text{dilation}} = v \cdot \Delta t = (0.900)(3.00 \times 10^8)(5.96 \times 10^{-8}) \approx 16.1\text{ m}$$

This factor-of-2.3 increase in travel distance is experimentally confirmed in particle physics detectors.

```python
# Visualise: proper time vs lab time
import numpy as np
import matplotlib.pyplot as plt

dt_tau = 2.60e-8   

beta_range = np.linspace(0.01, 0.9999, 2000)
gamma_range = 1 / np.sqrt(1 - beta_range**2)
dt_lab_range = gamma_range * dt_tau

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left: lab lifetime vs speed
axes[0].plot(beta_range, dt_lab_range * 1e8, color='steelblue', linewidth=2)
axes[0].axvline(0.900, color='crimson', linestyle='--', label=r'$v = 0.900c$ (this problem)')
axes[0].axhline(dt_lab_range[np.argmin(abs(beta_range - 0.9))] * 1e8,
                color='crimson', linestyle=':', alpha=0.6)
axes[0].scatter([0.900], [dt_lab_range[np.argmin(abs(beta_range - 0.9))] * 1e8],
                color='crimson', zorder=5, s=80)
axes[0].set_xlabel(r'$\beta = v/c$', fontsize=12)
axes[0].set_ylabel(r'Lab lifetime $\Delta t$ (×$10^{-8}$ s)', fontsize=12)
axes[0].set_title('Observed Pion Lifetime vs Speed', fontsize=13)
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

# Right: distance traveled vs speed
d_range = beta_range * 3e8 * dt_lab_range
d_classical = beta_range * 3e8 * dt_tau

axes[1].plot(beta_range, d_range, color='steelblue', linewidth=2, label='Relativistic distance')
axes[1].plot(beta_range, d_classical, color='orange', linewidth=2, linestyle='--', label='Classical (no dilation)')
axes[1].axvline(0.900, color='crimson', linestyle='--', alpha=0.7)
axes[1].set_xlabel(r'$\beta = v/c$', fontsize=12)
axes[1].set_ylabel('Distance traveled (m)', fontsize=12)
axes[1].set_title('Pion Travel Distance: Relativistic vs Classical', fontsize=13)
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)
axes[1].set_xlim(0, 1)

plt.tight_layout()
plt.show()
print("Plot saved.")
```

## Summary

$$\boxed{\Delta t = \gamma \,\Delta\tau \approx 2.294 \times 2.60 \times 10^{-8}\text{ s} = 5.96 \times 10^{-8}\text{ s}}$$

The laboratory observer sees the pion live nearly **2.3 times longer** than its intrinsic proper lifetime. This is a direct, experimentally verified consequence of time dilation — particles produced in accelerators and cosmic rays live long enough to be detected precisely because their internal clocks run slow from our perspective.
