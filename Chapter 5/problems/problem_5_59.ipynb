# Problem 5.59 — Relativistic Energy: Rest Energy of an Electron

## Problem Statement

What is the **rest energy** of an electron, given its mass is $m_e = 9.11 \times 10^{-31}$ kg? Give your answer **in joules and in MeV**.

---

## Background

Einstein's famous result from special relativity is the mass-energy equivalence:

$$E_0 = mc^2$$

where:
- $E_0$ is the **rest energy**
- $m$ is the **rest mass**
- $c = 3.00 \times 10^8$ m/s

The rest energy is an intrinsic property of the particle. It is the minimum energy the particle can have when its kinetic energy is zero.

### Unit Conversion: Joules to MeV

$$1\,\text{eV} = 1.602 \times 10^{-19}\,\text{J}$$
$$1\,\text{MeV} = 1.602 \times 10^{-13}\,\text{J}$$

---

## Given

| Quantity | Value |
|----------|-------|
| Electron rest mass | $m_e = 9.11 \times 10^{-31}$ kg |
| Speed of light | $c = 3.00 \times 10^8$ m/s |

```python
import math

# ── Given ─────────────────────────────────────────────────────────────
c   = 3.00e8       # speed of light, m/s
m_e = 9.11e-31     # rest mass of electron, kg

# ── Compute rest energy in Joules ─────────────────────────────────────
E0_joules = m_e * c**2

print("Rest Energy Calculation: E₀ = mc²")
print("=" * 50)
print(f"  m  = {m_e:.3e} kg")
print(f"  c  = {c:.3e} m/s")
print(f"  c² = {c**2:.4e} m²/s²")
print()
print(f"  E₀ = m × c²")
print(f"     = ({m_e:.3e}) × ({c**2:.4e})")
print(f"     = {E0_joules:.4e} J")
print()

# ── Convert to MeV ────────────────────────────────────────────────────
eV_per_joule  = 1 / 1.602e-19     # eV per joule
MeV_per_joule = eV_per_joule / 1e6

E0_eV  = E0_joules * eV_per_joule
E0_MeV = E0_joules * MeV_per_joule

print("Unit Conversion: Joules → eV → MeV")
print("=" * 50)
print(f"  1 eV  = 1.602 × 10⁻¹⁹ J")
print(f"  1 MeV = 1.602 × 10⁻¹³ J")
print()
print(f"  E₀ = {E0_joules:.4e} J × (1 eV / 1.602×10⁻¹⁹ J)")
print(f"     = {E0_eV:.4e} eV")
print(f"     = {E0_MeV:.4f} MeV")
print()
print("=" * 50)
print(f"ANSWERS:")
print(f"  E₀ = {E0_joules:.4e} J")
print(f"  E₀ = {E0_MeV:.4f} MeV  (≈ 0.511 MeV)")
print()
# Standard textbook value
print("Standard accepted value: E₀ = 0.511 MeV  ✓")
```

## Step-by-Step Solution

### 1. Apply the rest-energy formula

$$E_0 = m_e c^2 = (9.11 \times 10^{-31}\,\text{kg})(3.00 \times 10^8\,\text{m/s})^2$$

$$E_0 = (9.11 \times 10^{-31})(9.00 \times 10^{16})\,\text{J}$$

$$E_0 = 8.20 \times 10^{-14}\,\text{J}$$

### 2. Convert to MeV

$$E_0 = 8.20 \times 10^{-14}\,\text{J} \times \frac{1\,\text{MeV}}{1.602 \times 10^{-13}\,\text{J}}$$

$$\boxed{E_0 = 8.20 \times 10^{-14}\,\text{J} \approx 0.511\,\text{MeV}}$$


```python
import numpy as np
import matplotlib.pyplot as plt

c   = 3.00e8
m_e = 9.11e-31
E0  = m_e * c**2
MeV = 1.602e-13

# Total energy E = gamma * m_e * c^2 as a function of speed
beta_range  = np.linspace(0, 0.9999, 3000)
gamma_range = 1 / np.sqrt(1 - beta_range**2)

E_total_MeV = gamma_range * E0 / MeV      # total energy in MeV
K_MeV       = (gamma_range - 1) * E0 / MeV  # kinetic energy in MeV
E0_MeV_line = np.full_like(beta_range, E0 / MeV)  # rest energy

fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(beta_range, E_total_MeV, color='steelblue', linewidth=2, label=r'Total energy $E = \gamma m_e c^2$')
ax.plot(beta_range, K_MeV, color='darkorange', linewidth=2, linestyle='--', label=r'Kinetic energy $K = (\gamma-1)m_e c^2$')
ax.axhline(E0 / MeV, color='green', linewidth=1.5, linestyle=':', label=rf'Rest energy $E_0 = m_e c^2 = {E0/MeV:.3f}$ MeV')

ax.annotate(f'$E_0 = {E0/MeV:.3f}$ MeV', xy=(0.1, E0/MeV), xytext=(0.15, E0/MeV + 0.5), fontsize=11, color='green', arrowprops=dict(arrowstyle='->', color='green'))

ax.set_xlabel(r'eta = v/c', fontsize=13)
ax.set_ylabel('Energy (MeV)', fontsize=13)
ax.set_title(r'Electron Energies vs Speed ($E_0 = 0.511$ MeV)', fontsize=13)
ax.set_xlim(0, 1)
ax.set_ylim(0, 5)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)

plt.show()
```

## Summary

$$\boxed{E_0 = m_e c^2 = 8.20 \times 10^{-14}\,\text{J} = 0.511\,\text{MeV}}$$

The rest energy of an electron is **0.511 MeV** — one of the fundamental constants of particle physics. The graph shows how the total energy rises from this minimum value at rest, shooting up toward infinity as $v \to c$, while the kinetic energy starts at zero.
