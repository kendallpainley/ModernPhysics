## Relativistic Television

A non-flat screen, older-style television display works by accelerating electrons over a short
distance to relativistic speed, and then using electromagnetic fields to control where the electron beam strikes a
fluorescent layer at the front of the tube. Suppose the electrons travel at $6.00 × 107 m/s$ through a distance of
$0.200 m$ from the start of the beam to the screen.
- What is the time of travel of an electron in the rest frame of the television set?
- What is the electron’s time of travel in its own rest frame?

## Strategy

**For a:**
- We are working entirely within the TV set's frame of reference
- Electrons velocity: $6.00 × 10^7$ m/s
- Distance from the start of the beam to the screen: $0.200$ m
- We want to find the time, $\Delta t$, from $vt = d$

**Part (a): Time in the TV set's frame**

**Step 1: Write down the formula**

Use the classical relationship between distance, speed, and time:

$$\Delta t = \frac{d}{v}$$

**Step 2: Plug in the numbers**

$$\Delta t = \frac{0.200\ \text{m}}{6.00 \times 10^7\ \text{m/s}}$$

$$\boxed{\Delta t = 3.33 \times 10^{-9}\ \text{s}}$$

### Strategy

**For b:**

The two events (the electron leaving the gun, and the screen arriving at the electron) both happen at the same location in the electron's frame. That makes the time between them the proper time $Δτ$.

- Electrons velocity: $6.00 × 10^7$ m/s
- Distance from the start of the beam to the screen: $0.200$ m
- Dilated time $Δt$: $3.33 × 10⁻⁹$ s
- We want to relate the proper time $\Delta \tau $ in the electron rest frame to the time in the television frame $\Delta t$

**Part (b): Time in the electron's own frame**

**Step 1: Write down the formula**

$$\Delta t = \frac{\Delta\tau}{\sqrt{1 - v^2/c^2}} \implies \Delta\tau = \Delta t \sqrt{1 - v^2/c^2}$$

**Step 2: Calculate the square root factor**

$$\frac{v}{c} = \frac{6.00 \times 10^7}{3.00 \times 10^8} = 0.200$$

$$\left(\frac{v}{c}\right)^2 = 0.200^2 = 0.0400$$

$$1 - 0.0400 = 0.9600$$

$$\sqrt{0.9600} = 0.9798$$

**Step 3: Multiply to find proper time**

$$\Delta\tau = (3.33 \times 10^{-9}\ \text{s})(0.9798)$$

$$\boxed{\Delta\tau = 3.26 \times 10^{-9}\ \text{s}}$$
