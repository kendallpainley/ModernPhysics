### Calculating a Doppler Shift
A galaxy is moving away from Earth at $v = 0.825c$. It emits radio waves with a wavelength of $\lambda_s = 0.525$ m in its own rest frame. What wavelength do we detect on Earth?

#### Strategy
Because the galaxy is moving at a relativistic speed, we use the **relativistic Doppler formula** rather than the classical one. Since the galaxy is moving **away**, we expect the observed wavelength to be longer than the emitted wavelength — a **red shift**.

Known values:
- $v = 0.825c$ (recession speed, positive = moving away)
- $\lambda_s = 0.525$ m (wavelength emitted in galaxy's rest frame)

**Step 1: Write the relativistic Doppler formula**

$$\lambda_\text{obs} = \lambda_s \sqrt{\frac{1 + v/c}{1 - v/c}}$$

where $v$ is positive when the source is moving away from the observer.

---

**Step 2: Evaluate the fraction under the square root**

$$\frac{v}{c} = 0.825$$

$$\frac{1 + 0.825}{1 - 0.825} = \frac{1.825}{0.175} = 10.43$$

$$\sqrt{10.43} = 3.23$$

---

**Step 3: Multiply by the source wavelength**

$$\lambda_\text{obs} = 0.525\ \text{m} \times 3.23$$

$$\boxed{\lambda_\text{obs} = 1.70\ \text{m}}$$

The galaxy emits at 0.525 m but we receive at 1.70 m. This large red shift is caused by the galaxy receding at 82.5% of the speed of light.
