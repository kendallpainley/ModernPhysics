### Comparing Kinetic Energy
An electron travels at $v = 0.990c$. The mass of an electron is $m = 9.11 \times 10^{-31}$ kg.

**(a)** Calculate the electron's relativistic kinetic energy in MeV.

**(b)** Calculate the classical kinetic energy at the same speed and compare the two results.

#### Strategy
At 99% of the speed of light, relativistic effects are very large. We calculate $\gamma$ first, use it for the relativistic kinetic energy, then compare with $K_\text{class} = \frac{1}{2}mv^2$.

Known values:
- $v = 0.990c$
- $m = 9.11 \times 10^{-31}$ kg
- $c = 3.00 \times 10^8$ m/s
- $1\ \text{MeV} = 1.60 \times 10^{-13}$ J

#### Part (a): Relativistic kinetic energy

**Step 1: Write the formula**

$$K_\text{rel} = (\gamma - 1)mc^2 \quad \text{where} \quad \gamma = \frac{1}{\sqrt{1 - v^2/c^2}}$$

**Step 2: Calculate $\gamma$**

$$\frac{v^2}{c^2} = (0.990)^2 = 0.9801$$

$$\sqrt{1 - 0.9801} = \sqrt{0.0199} = 0.1411$$

$$\gamma = \frac{1}{0.1411} = 7.089$$

**Step 3: Calculate kinetic energy in joules**

$$K_\text{rel} = (7.089 - 1)(9.11 \times 10^{-31})(3.00 \times 10^8)^2$$

$$= (6.089)(9.11 \times 10^{-31})(9.00 \times 10^{16})$$

$$= 4.99 \times 10^{-13}\ \text{J}$$

**Step 4: Convert to MeV**

$$K_\text{rel} = \frac{4.99 \times 10^{-13}}{1.60 \times 10^{-13}}$$

$$\boxed{K_\text{rel} = 3.12\ \text{MeV}}$$

---

#### Part (b): Classical kinetic energy

$$K_\text{class} = \frac{1}{2}mv^2 = \frac{1}{2}(9.11 \times 10^{-31})(0.990 \times 3.00 \times 10^8)^2$$

$$= \frac{1}{2}(9.11 \times 10^{-31})(8.83 \times 10^{16}) = 4.02 \times 10^{-14}\ \text{J}$$

$$K_\text{class} = \frac{4.02 \times 10^{-14}}{1.60 \times 10^{-13}} = \boxed{0.251\ \text{MeV}}$$

---

#### Comparison

| Formula | Kinetic Energy |
|---|---|
| Relativistic | 3.12 MeV |
| Classical | 0.251 MeV |
| Ratio $K_\text{rel}/K_\text{class}$ | **12.4** |

The classical formula underestimates the kinetic energy by a factor of 12.4 at this speed. As $v \to c$, $\gamma \to \infty$, meaning an infinite amount of energy would be required to reach the speed of light — confirming that no massive object can ever attain it.
