## Calculating Length Contraction

Suppose an astronaut travels so fast that $γ = 30.00$.
- (a) The astronaut travels from Earth to the nearest star system, Alpha Centauri, 4.300 light years (ly) away as measured
by an earthbound observer. How far apart are Earth and Alpha Centauri as measured by the astronaut?
- (b) In terms of $c$, what is the astronaut’s velocity relative to Earth? (You may neglect the motion of Earth relative to the sun)

### Strategy

For part a:
- Distance between Earth and Alpha Centauri, measured from earth-observer: $L_0$ = $4.300 ly$
- Lorentz Factor: $γ = 30.00$


**Part (a): The contracted distance**

**Step 1: Write down the formula**

The astronaut is in motion relative to both endpoints, so the distance between them is length-contracted. We use the length contraction formula:

$$L = \frac{L_0}{\gamma}$$

**Step 2: Plug in the numbers**

$$L = \frac{4.300\ \text{ly}}{30.00}$$

$$\boxed{L = 0.1433\ \text{ly}}$$

The astronaut sees the distance to Alpha Centauri as only about $0.143$ light-years.

### Strategy

For part b:
- Lorentz Factor: $γ = 30.00$
- Find $v$ by rearranging $γ$ to express $v$ in terms of $c$

**Part (b): Finding the speed from $γ$**

**Step 1: Write down the formula**

$$\gamma = \frac{1}{\sqrt{1 - v^2/c^2}}$$

To solve this for $v/c$, start by squaring both sides:

$$\gamma^2 = \frac{1}{1 - v^2/c^2}$$

Rearrange to isolate $v^2/c^2$:

$$1 - \frac{v^2}{c^2} = \frac{1}{\gamma^2}$$

$$\frac{v^2}{c^2} = 1 - \frac{1}{\gamma^2}$$

Take the square root:

$$\frac{v}{c} = \sqrt{1 - \frac{1}{\gamma^2}}$$

**Step 2: Plug in $γ = 30.00$**

$$\frac{v}{c} = \sqrt{1 - \frac{1}{(30.00)^2}} = \sqrt{1 - \frac{1}{900}} = \sqrt{1 - 0.001111} = \sqrt{0.998889}$$

$$\frac{v}{c} = 0.99944$$

$$\boxed{v = 0.99944c}$$

The astronaut is travelling at 99.944% of the speed of light.
