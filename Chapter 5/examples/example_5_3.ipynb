## Calculating $Δt$ for a Relativistic Event
Suppose a cosmic ray colliding with a nucleus in Earth’s upper atmosphere produces a muon that has a velocity
$v = 0.950c$. The muon then travels at constant velocity and lives $2.20 μs$ as measured in the muon’s frame of
reference.How long does the muon live, as measured by an earthbound observer?

### Strategy

- A clock moving with the muon measures the proper time of its decay process: $Δτ = 2.20μs$
- Muon velocity: $v = 0.950c$
- We want to find the earthbound observer's measure of $Δt$

$**Step 1: Write down the formula**

Start with the time dilation formula:

$\Delta t = \gamma \Delta\tau = \frac{\Delta\tau}{\sqrt{1 - v^2/c^2}}$

---

**Step 2: Calculate γ**

$$\gamma = \frac{1}{\sqrt{1 - v^2/c^2}}$$

The muon travels at $v = 0.950c$, so:

$$\frac{v^2}{c^2} = (0.950)^2 = 0.9025$$

$$1 - 0.9025 = 0.0975$$

$$\sqrt{0.0975} = 0.3122$$

$$\gamma = \frac{1}{0.3122} = 3.20$$

---

**Step 3: Multiply by the proper time**

$$\Delta t = \gamma \Delta\tau = 3.20 \times 2.20\ \mu\text{s}$$

$$\boxed{\Delta t = 7.05\ \mu\text{s}}$$

---

The muon's own clock ticks off $2.20 μs$ before it decays. But from Earth's point of view, that same muon lives for $7.05 μs$.

A muon travelling at $0.950c$ covers roughly $2$ km in $7.05 μs$ as seen from Earth. Without time dilation, its $2.20 μs$ lifetime would only carry it about $0.6$ km — not nearly far enough to reach the ground from the upper atmosphere where it was created. Yet muons are detected at Earth's surface all the time. Their extended apparent lifetime, caused by time dilation, is what allows them to complete the journey.
