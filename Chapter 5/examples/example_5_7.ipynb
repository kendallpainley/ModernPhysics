### Using the Lorentz Transformation for Length
A surveyor on Earth measures a street to be $L = 100$ m long. A spaceship S′ passes by at $v = 0.20c$, moving parallel to the street. Use the Lorentz transformation to find the length of the street as measured by the spaceship crew.

#### Strategy
The surveyor in frame S measures both ends of the street **simultaneously** at the same time $t$, finding separation $L = x_2 - x_1 = 100$ m. The spaceship crew (frame S′) measures the same two endpoints in their frame. We apply the Lorentz position transformation to relate the two measurements.

Known values:
- $L = x_2 - x_1 = 100$ m (proper length, measured in Earth's frame)
- $v = 0.20c$
- $\Delta\tau = 0$ (both endpoints measured simultaneously in S)

**Step 1: Apply the Lorentz position transformation to both endpoints**

For both endpoints measured at the same time $t$ in S:

$$x'_2 - x'_1 = \frac{(x_2 - vt) - (x_1 - vt)}{\sqrt{1 - v^2/c^2}} = \frac{x_2 - x_1}{\sqrt{1 - v^2/c^2}} =  \frac{L}{\sqrt{1 - v^2/c^2}}$$


**Step 2: Calculate the contracted length**

Because $x'_2 − x'_1 = 100$ m, the length of the moving stick is equal to:

$$L′ = (100 m) \sqrt{1 - v^2/c^2}$$

$$L' = (100 m) \sqrt{1 - (0.20)^2}$$

$$\boxed{L' = 98.0\ \text{m}}$$

The street appears 98.0 m long to the spaceship crew. The Lorentz transformation automatically reproduces the length contraction formula, confirming that length contraction is a direct consequence of how coordinates transform between frames.
