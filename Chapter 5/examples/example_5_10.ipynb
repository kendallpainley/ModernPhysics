### Relativistic Package Delivery
A spaceship approaches Earth at $v = 0.500c$ and fires a canister at $u' = 0.750c$ relative to the ship.

**(a)** What is the canister's speed as seen from Earth if fired **toward** Earth?

**(b)** What is the canister's speed as seen from Earth if fired **away from** Earth?

#### Strategy
Use relativistic velocity addition. The sign of $u'$ indicates direction: positive toward Earth, negative away from Earth.

$$u = \frac{u' + v}{1 + \dfrac{vu'}{c^2}}$$

Known values:
- $v = 0.500c$ (ship's speed toward Earth)
- $u' = \pm 0.750c$ (canister speed in ship's frame)

#### Part (a): Canister fired toward Earth ($u' = +0.750c$)

**Numerator:**
$$u' + v = 0.750c + 0.500c = 1.250c$$

**Denominator:**
$$1 + \frac{(0.500c)(0.750c)}{c^2} = 1 + 0.375 = 1.375$$

**Result:**
$$u = \frac{1.250c}{1.375}$$

$$\boxed{u = 0.909c}$$

---

#### Part (b): Canister fired away from Earth ($u' = -0.750c$)

**Numerator:**
$$u' + v = -0.750c + 0.500c = -0.250c$$

**Denominator:**
$$1 + \frac{(0.500c)(-0.750c)}{c^2} = 1 - 0.375 = 0.625$$

**Result:**
$$u = \frac{-0.250c}{0.625}$$

$$\boxed{u = -0.400c}$$

The negative sign confirms the canister moves away from Earth.
