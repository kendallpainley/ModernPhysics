### Using the Lorentz Transformation for Time
Spacecraft S′ passes spacecraft S at a relative speed of $c/2$. The captain of S′ sends a radio signal that lasts $\Delta t' = 1.2$ s according to S′'s own clock. Use the Lorentz transformation to find the time interval of the signal as measured by spacecraft S.

#### Strategy
The signal is emitted from a fixed location aboard S′, so both the start and end of the signal happen at the **same place** in S′. This means $\Delta x' = 0$, and the Lorentz transformation simplifies to the time dilation formula.

Known values:
- $\Delta t' = 1.2$ s
- $\Delta x' = 0$ (signal emitted from a fixed clock aboard S′)
- $v = c/2$

**Step 1: Write the Lorentz transformation for time**

$$\Delta t = \frac{\Delta t' + v\Delta x'/c^2}{\sqrt{1 - v^2/c^2}}$$

---

**Step 2: Apply $\Delta x' = 0$**

Since the signal is emitted from one fixed location aboard S′, $\Delta x' = 0$ and the equation simplifies to:

$$\Delta t = \frac{\Delta t'}{\sqrt{1 - v^2/c^2}}$$


---

**Step 3: Calculate**

$$v = \frac{c}{2},\quad \frac{v^2}{c^2} = \frac{1}{4} = 0.25$$

$$\sqrt{1 - 0.25} = \sqrt{0.75} = 0.8660$$

$$\Delta t = \frac{1.2\ \text{s}}{0.8660}$$

$$\boxed{\Delta t = 1.6\ \text{s}}$$

The signal lasts 1.2 s aboard S′ but 1.6 s as measured by S — a factor of $\gamma \approx 1.155$ at $v = 0.5c$. This confirms that the Lorentz transformation is the more general tool: time dilation is simply a special case of it, which applies whenever the two events occur at the same location in the moving frame.
