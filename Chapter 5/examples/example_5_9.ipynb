### Velocity Transformation Equations for Light
A spaceship heading directly toward Earth at $v = 0.500c$ fires a laser beam toward Earth. The laser light leaves the ship at speed $u' = c$ as observed from the ship. Calculate the speed at which the laser beam approaches Earth.

#### Strategy
Because the spaceship and the light are both moving at relativistic speeds, we cannot use simple velocity addition. We use the relativistic velocity addition formula instead.

Known values:
- $v = 0.500c$ (speed of ship relative to Earth)
- $u' = c$ (speed of laser in ship's frame)

**Step 1: Write the relativistic velocity addition formula**

$$u = \frac{u' + v}{1 + \dfrac{vu'}{c^2}}$$

---

**Step 2: Substitute the values**

**Numerator:**
$$u' + v = c + 0.500c = 1.500c$$

**Denominator:**
$$1 + \frac{(0.500c)(c)}{c^2} = 1 + 0.500 = 1.500$$

**Result:**
$$u = \frac{1.500c}{1.500}$$

$$\boxed{u = c}$$

The laser beam reaches Earth at exactly the speed of light. This is not a coincidence: the relativistic velocity addition formula is specifically constructed so that whenever $u' = c$, the result is always $c$ regardless of $v$. This is Einstein's second postulate in action.
