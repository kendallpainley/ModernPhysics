### Time Dilation in a High-Speed Vehicle
The Hypersonic Technology Vehicle 2 (HTV-2) is an experimental rocket vehicle capable of traveling at $21,000
 \frac{km}{h} (5830 \frac{m}{s})$. If an electronic clock in the HTV-2 measures a time interval of exactly 1-s duration, what would
observers on Earth measure the time interval to be?

#### Strategy
Apply the time dilation formula to relate the proper time interval of the signal in HTV-2 to the time interval
measured on the ground.

- The clock inside the vehicle is measuring its own time, the proper time, $Δτ = 1 s$.<br>
- The speed of the vehicle is $v = 5830 \frac{m}{s}$.<br>
- We want $Δt$, the Earth observer's measurement.

**Step 1: Write down the formula**

From the time dilation derivation, we know that a moving clock ticks slower than a stationary one. The relationship between the two time measurements is:

$$\Delta t = \frac{\Delta\tau}{\sqrt{1 - v^2/c^2}}$$

Where:
- $\Delta\tau$ is the **proper time**
- $\Delta t$ is the **dilated time**
- $v$ is the speed of the vehicle
- $c$ is the speed of light: $3.00 \times 10^8$ m/s

---

**Step 2: Work out the fraction inside the square root**

Before we can evaluate the square root, we need to calculate $v^2/c^2$ — which tells us the vehicle's speed as a proportion of the speed of light, squared.

First, find $v/c$:

$$\frac{v}{c} = \frac{5830 \text{ m/s}}{3.00 \times 10^8 \text{ m/s}} = 0.0000194$$


Now square that fraction to get $v^2/c^2$:

$$\left(\frac{v}{c}\right)^2 = (0.0000194)^2 = 3.77 \times 10^{-10}$$

---

**Step 3: Subtract from 1**

Now evaluate the expression inside the square root:

$$1 - \left(\frac{v}{c}\right)^2 = 1 - 0.000000000377 = 0.999999999623$$

---

**Step 4: Take the square root**

Next take the square root of that result:

$$\sqrt{0.999999999623} \approx 0.999999999811$$

---

**Step 5: Divide Δτ by that number**

Now we divide the proper time by the square root we just calculated:

$$\Delta t = \frac{1 \text{ s}}{0.999999999811} = 1.000000000189 \text{ s}$$

$$\boxed{\Delta t = 1 \text{ s} + 1.89 \times 10^{-10} \text{ s}}$$

So the Earth observer measures the same event lasting **189 picoseconds longer** than the vehicle's own clock recorded.
