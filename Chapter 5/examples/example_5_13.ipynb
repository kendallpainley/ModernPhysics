### Calculating Rest Energy
Calculate the rest energy of a $m = 1.00$ g mass.

#### Strategy
A stationary object contains energy simply by virtue of having mass. We multiply the mass by $c^2$ to find the equivalent rest energy.

Known values:
- $m = 1.00\ \text{g} = 1.00 \times 10^{-3}$ kg
- $c = 3.00 \times 10^8$ m/s

**Step 1: Write down the formula**

$$E_0 = mc^2$$


---

**Step 2: Calculate**

$$E_0 = (1.00 \times 10^{-3}\ \text{kg})(3.00 \times 10^8\ \text{m/s})^2$$

$$= (1.00 \times 10^{-3})(9.00 \times 10^{16}\ \text{m}^2/\text{s}^2)$$

$$\boxed{E_0 = 9.00 \times 10^{13}\ \text{J}}$$

---
