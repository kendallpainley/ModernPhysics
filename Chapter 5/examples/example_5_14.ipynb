### Calculating Rest Mass
A car battery is rated at $600$ ampere-hours (A·h) at $12.0$ V. The battery's total mass is $20.0$ kg.

**(a)** By how much does the battery's rest mass increase when it is fully charged (assuming no chemical reactants enter or leave)?

**(b)** What percentage of the battery's total mass does this increase represent?

#### Strategy
Stored energy has mass — a charged battery literally weighs more than a depleted one. We first calculate the electrical energy stored in the battery, then use $E = \Delta m \cdot c^2$ to find the corresponding mass increase.

Known values:
- $I \cdot t = 600$ A·h
- $V = 12.0$ V
- $m = 20.0$ kg
- $c = 3.00 \times 10^8$ m/s

#### Part (a): Mass increase

**Step 1: Calculate the total charge in coulombs**

One ampere = one coulomb per second, and one hour = 3600 seconds:

$$q = 600\ \text{A·h} \times \frac{3600\ \text{s}}{1\ \text{h}} = 2.16 \times 10^6\ \text{C}$$

**Step 2: Calculate the energy stored**

$$E = qV = (2.16 \times 10^6\ \text{C})(12.0\ \text{V}) = 2.59 \times 10^7\ \text{J}$$

**Step 3: Convert energy to mass**

From $E = \Delta m \cdot c^2$, rearranging:

$$\Delta m = \frac{E}{c^2} = \frac{2.59 \times 10^7\ \text{J}}{(3.00 \times 10^8\ \text{m/s})^2} = \frac{2.59 \times 10^7}{9.00 \times 10^{16}}$$

$$\boxed{\Delta m = 2.88 \times 10^{-10}\ \text{kg}}$$


---

#### Part (b): Percentage increase

$$\%\ \text{increase} = \frac{\Delta m}{m} \times 100\% = \frac{2.88 \times 10^{-10}\ \text{kg}}{20.0\ \text{kg}} \times 100\%$$

$$\boxed{\%\ \text{increase} = 1.44 \times 10^{-9}\%}$$

That is roughly one part in 70 billion — completely undetectable by any scale. This is why the law of conservation of mass appeared exact for centuries: chemical and electrical processes store or release energy, but the corresponding mass change is too small to measure. Only in nuclear reactions, where a far larger fraction of mass is converted, does the mass change become observable. A charged battery really does weigh more than a dead one — we simply cannot measure the difference.
