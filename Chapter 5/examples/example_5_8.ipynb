### Lorentz Transformation and Simultaneity
An observer standing by the railway tracks sees two light bulbs flash **simultaneously** at both ends of a 26 m long passenger car when the middle of the car passes him at $v = c/2$. A passenger sits in the middle of the car. Find the time separation between the two flashes as measured by the passenger.

#### Strategy
In the ground frame S, the two flashes happen at the same time: $\Delta t = 0$. The spatial separation between them is the length of the car: $\Delta x' = 26$ m (in the passenger's frame, this is the proper length of the car). We use the Lorentz transformation to find $\Delta t'$, the time between flashes in the passenger's frame.

Known values:
- $\Delta t = 0$ (simultaneous in the ground frame)
- $\Delta x' = 26$ m (proper length of the car)
- $v = c/2$

**Step 1: Write the Lorentz transformation for time**

$$\Delta t = \frac{\Delta t' + v\Delta x'/c^2}{\sqrt{1 - v^2/c^2}}$$

---

**Step 2: Set $\Delta t = 0$ and solve for $\Delta t'$**

$$0 = \frac{\Delta t' + \frac{v}{c^2}\Delta x'}{\sqrt{1 - v^2/c^2}}$$

For the numerator to equal zero:

$$\Delta t' = -\frac{v \cdot \Delta x'}{c^2}$$

---

**Step 3: Calculate**

$$\Delta t' = -\frac{\frac{c}{2} \times 26\ \text{m}}{c^2} = -\frac{26\ \text{m}}{2c}$$

$$\Delta t' = -\frac{26\ \text{m}}{2 \times 3.00 \times 10^8\ \text{m/s}}$$

$$\boxed{\Delta t' = -4.33 \times 10^{-8}\ \text{s}}$$

The passenger does **not** see the flashes as simultaneous — she measures them 43.3 nanoseconds apart. The negative sign tells us the right-end flash happens **first** in her frame: the car is moving to the right, so she is approaching the right-side flash and the light has less distance to travel to reach her. This is a direct numerical demonstration that **simultaneity is not absolute** — two events that are simultaneous in one frame need not be simultaneous in another.
