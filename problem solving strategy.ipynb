### Problem-Solving Strategy

1. Make a list of what is given or can be inferred from the problem as stated.
2. Identify exactly what needs to be determined in the problem.
3. Make certain to understand the conceptual aspects of the problem before making any calculations. Decide, for example, which observer sees time dilated or length contracted before working with the equations or using them to carry out the calculation.
4. Determine the primary type of calculation to be done to find the unknowns identified.
